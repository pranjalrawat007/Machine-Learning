```python
import pandas as pd
from scipy.optimize import linprog
data = pd.read_csv('data/stigler.csv')
nutrients = [
    ["calories", 3], ["protein_g", 70], ["calcium_g", 0.8], ["iron_mg", 12], 
    ["vitamin_a_iu", 5], ["vitamin_b1_mg", 1.8], ["vitamin_b2_mg", 2.7], 
    ["niacin_mg", 18], ["vitamin_c_mg", 75]
]
c = data['price_cents'].to_numpy() / 100
A = -data.iloc[:, 3:].to_numpy().T  # Transpose to match nutrients vs. foods
b = -pd.DataFrame(nutrients)[1].to_numpy()
c = data['price_cents'].to_numpy() / 100
bounds = [(0, None) for _ in range(len(c))]
result = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')

# Display the results
if result.success:
    total_cost = 365 * result.fun
    print("\nAnnual cost: ${:.2f}".format(total_cost))
    print("\nAnnual foods:")
    for i, quantity in enumerate(result.x):
        if quantity > 0:
            print(f"{data['commodity'][i]}: {quantity * 365:.2f} units/year")

    print("\nNutrients per day:")
    for i, nutrient in enumerate(nutrients):
        nutrient_total = sum(data.iloc[j, i + 3] * result.x[j] for j in range(len(result.x)))
        print(f"{nutrient[0]}: {nutrient_total:.2f} (min {nutrient[1]})")
else:
    print("No optimal solution found.")
```

```python
Ax = np.dot(A, result.fun)
print(result.fun, result.x, Ax)
```

```python
import numpy as np
from scipy.optimize import linprog
C = np.array([[8, 6, 10], [9, 7, 4]])
s = np.array([20, 30])
d = np.array([15, 25, 10])
c = C.flatten()
A_eq_supply = np.kron(np.eye(len(s)), np.ones(len(d)))
A_eq_demand = np.kron(np.ones(len(s)), np.eye(len(d)))
A_eq = np.vstack([A_eq_supply, A_eq_demand])
b_eq = np.concatenate([s, d])
bounds = [(0, None) for _ in range(len(c))]
result = linprog(c, A_eq=A_eq, b_eq=b_eq, bounds=bounds)
print(result.x.reshape(C.shape), result.fun)
```
