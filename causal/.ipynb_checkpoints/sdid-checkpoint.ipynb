### DID

- Use FE: To control for group-specific unobserved variables.
- Use Clustered SEs: To address correlation in errors within groups.
- Use Both: When both omitted variable bias and within-group correlation are concerns. Particularly important in clustered sample designs or when heterogeneity in treatment effects exists.

```python
import numpy as np
import pandas as pd
from scipy.optimize import minimize
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

data = pd.read_csv('data/california99.csv')
data['T_it'] = ((data['state'] == 'California') & (data['year'] >= 1989)).astype(int)
prepared_data = data.rename(columns={'state': 'i', 'year': 't', 'cigsale': 'Y_it', 'lnincome': 'X1_it', 'beer': 'X2_it', 'age15to24': 'X3_it', 'retprice': 'X4_it'})

prepared_data['location'] = (prepared_data['i'] == 'California').astype(int)
prepared_data['period'] = (prepared_data['t'] >= 1989).astype(int)
prepared_data['location_period'] = prepared_data['location'] * prepared_data['period']

Y = prepared_data['Y_it'].values
D = prepared_data['T_it'].values
units = prepared_data['i'].astype('category').cat.codes.values
time = prepared_data['t'].astype('category').cat.codes.values
N = len(np.unique(units))
T = len(np.unique(time))

Y_bar_t = np.array([Y[time == t].mean() for t in range(T)])
D_bar_t = np.array([D[time == t].mean() for t in range(T)])
Y_bar_i = np.array([Y[units == i].mean() for i in range(N)])
D_bar_i = np.array([D[units == i].mean() for i in range(N)])

Y_demeaned = Y - Y_bar_i[units] - Y_bar_t[time] + Y.mean()
D_demeaned = D - D_bar_i[units] - D_bar_t[time] + D.mean()

def objective(params):
    tau = params[0]
    residual = Y_demeaned - tau * D_demeaned
    return np.sum(residual**2)

result = minimize(objective, [0.0], method='BFGS')
manual_tau_hat = result.x[0]

model = smf.ols(formula='Y_it ~ location + period + location_period', data=prepared_data).fit(cov_type='cluster', cov_kwds={'groups': prepared_data['i']})
regression_tau_hat = model.params['location_period']

pre_period = prepared_data[prepared_data['t'] < 1989]
post_period = prepared_data[prepared_data['t'] >= 1989]

mean_ca_pre = pre_period[pre_period['i'] == 'California']['Y_it'].mean()
mean_non_ca_pre = pre_period[pre_period['i'] != 'California']['Y_it'].mean()
mean_ca_post = post_period[post_period['i'] == 'California']['Y_it'].mean()
mean_non_ca_post = post_period[post_period['i'] != 'California']['Y_it'].mean()

adjusted_ca_post = mean_ca_pre + regression_tau_hat
print(manual_tau_hat, regression_tau_hat)
plt.figure(figsize=(8, 6))
plt.plot([1988, 1989], [mean_ca_pre, mean_ca_post], label='California', color='blue', marker='o', linestyle='-')
plt.plot([1988, 1989], [mean_non_ca_pre, mean_non_ca_post], label='Control (Observed)', color='red', marker='o', linestyle='-')
plt.plot([1988, 1989], [mean_ca_pre, adjusted_ca_post], label='Adjusted (Parallel to Control)', color='green', linestyle='--', marker='o')
plt.axvline(x=1989, color='black', linestyle=':', label='Proposition 99')
plt.xlabel('Year')
plt.ylabel('Cigarette Sales')
plt.title('DiD Estimation with Regression Adjustment')
plt.legend()
plt.grid(True)
plt.show()

import matplotlib.pyplot as plt

# Prepare data for plotting
pivot_df = prepared_data.pivot_table(index='t', columns='i', values='Y_it')
california_sales = pivot_df['California']
non_california_sales = pivot_df.drop('California', axis=1).mean(axis=1)

# Plotting pre-treatment trends
plt.figure(figsize=(10, 5))
plt.plot(non_california_sales.index, non_california_sales.values, color="C1", alpha=0.3, label="Control")
plt.plot(california_sales.index, california_sales.values, color="C0", label="California")
plt.axvline(x=1989, ymin=40, ymax=300, linestyle=":", lw=2, label="Proposition 99", color="black")
plt.legend()
plt.ylabel("Cigarette Sales")
plt.title("Non-Parallel Trends")
plt.show()
```

### Synthetic Control

```python
# Reload the base dataset
data = pd.read_csv(file_path)

# Step 1: Pre-treatment Data Preparation
# Define treated and post indicators
data['treated'] = (data['state'] == 'California').astype(int)
data['post'] = (data['year'] >= 1989).astype(int)

# Extract pre-period data
pre_period_data = data[data['post'] == 0]

# Pivot pre-period data to create matrices for synthetic control
pre_pivoted_data = pre_period_data.pivot(index='year', columns='state', values='cigsale')
pre_pivoted_data = pre_pivoted_data[[col for col in pre_pivoted_data.columns if col != 'California'] + ['California']]

# Define pre-treatment matrices
Y_pre_control = pre_pivoted_data.iloc[:, :-1].values  # Control units
Y_pre_treatment = pre_pivoted_data.iloc[:, -1].values  # California

# Step 2: Synthetic Control Weights Calculation
# Define regularization term
T_pre = Y_pre_control.shape[0]
M = Y_pre_control.shape[1]
eta = (T_pre * M) ** (1 / 4) * np.std(np.diff(Y_pre_treatment))

# Define the objective function for weight calculation
def synthetic_control_objective(w):
    w0 = w[-1]  # Intercept
    weights = w[:-1]
    fit = Y_pre_control @ weights + w0
    penalty = eta * T_pre * np.sum(weights**2)  # Regularization
    return np.sum((Y_pre_treatment - fit)**2) + penalty

# Constraints: weights >= 0 and sum(weights) = 1
constraints = [
    {'type': 'eq', 'fun': lambda w: np.sum(w[:-1]) - 1},  # Sum of weights = 1
    {'type': 'ineq', 'fun': lambda w: w[:-1]}             # Weights >= 0
]

# Initial guess: equal weights and zero intercept
initial_guess = np.zeros(M + 1)
initial_guess[:-1] = 1 / M

# Perform optimization
result = minimize(synthetic_control_objective, initial_guess, constraints=constraints)
optimized_weights = result.x[:-1]
optimized_intercept = result.x[-1]

# Identify states with significant weights
significant_weights = dict(zip(pre_pivoted_data.columns[:-1], optimized_weights))

# Step 3: Post-treatment Prediction
# Extract post-treatment data for control units
post_period_data = data[data['post'] == 1]
post_pivoted_data = post_period_data.pivot(index='year', columns='state', values='cigsale')
post_pivoted_data = post_pivoted_data[[col for col in post_pivoted_data.columns if col != 'California']]

# Predict California's post-treatment outcomes
Y_post_control = post_pivoted_data.values
Y_predicted_california = Y_post_control @ optimized_weights + optimized_intercept

# Assign predictions to the dataset
data['Y_synth_control'] = np.nan
data.loc[data['treated'] == 1, 'Y_synth_control'] = np.concatenate([Y_pre_control @ optimized_weights + optimized_intercept, Y_predicted_california])

# Step 4: Compute and Plot ATE_SC
california_data = data[data['treated'] == 1].copy()
california_data['ATE_SC'] = california_data['Y_synth_control'] - california_data['cigsale']
print(california_data[california_data.post  == 1].ATE_SC.mean())
      
# Plot ATE_SC over time
plt.figure(figsize=(10, 6))
plt.plot(california_data['year'], california_data['ATE_SC'], label='ATE_SC', marker='o')
plt.axvline(x=1989, color='red', linestyle=':', label='1989 (Treatment Start)')
plt.xlabel('Year')
plt.ylabel('ATE_SC (Y_synth_control - Y_actual_california)')
plt.title('ATE_SC Over Time')
plt.legend()
plt.grid(True)
plt.show()

# Plot California's actual outcomes vs. synthetic control predictions
plt.figure(figsize=(10, 6))

# Plot actual California outcomes
plt.plot(california_data['year'], california_data['cigsale'], label='Actual California', marker='o')

# Plot synthetic control predictions
plt.plot(california_data['year'], california_data['Y_synth_control'], label='Synthetic Control', linestyle='--', marker='x')

# Add a vertical line for the year 1989
plt.axvline(x=1989, color='red', linestyle=':', label='1989 (Treatment Start)')

# Add labels, legend, and title
plt.xlabel('Year')
plt.ylabel('Cigarette Sales')
plt.title('Actual California vs. Synthetic Control Over Time')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

### Time Weights

```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize
from numpy.linalg import norm

# Load the dataset
file_path = 'data/california99.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Step 1: Define treated and post indicators
data['treated'] = (data['state'] == 'California').astype(int)
data['post'] = (data['year'] >= 1989).astype(int)

# Step 2: Separate control units
control_data = data[data['treated'] == 0]

# Step 3: Pre-treatment Data Preparation
pre_period_data = control_data[control_data['post'] == 0]
pre_pivoted_data = pre_period_data.pivot(index='year', columns='state', values='cigsale')

# Step 4: Post-treatment Data Preparation
post_period_data = control_data[control_data['post'] == 1]
Y_avg_control_in_post = post_period_data.groupby('state')['cigsale'].mean().values  # Mx1 vector

# Step 5: Define Y_pre_control (T_pre x M matrix of pre-treatment data for controls)
Y_pre_control = pre_pivoted_data.values  # T_pre x M
Y_pre_control_transposed = Y_pre_control.T  # Transpose to align with MxT_pre

# Step 6: Define the objective function for lambda optimization
def lambda_objective(l):
    lambda_weights = l[:-1]  # Lambda weights (excluding the intercept)
    lambda0 = l[-1]  # Intercept term
    fit = Y_pre_control_transposed @ lambda_weights + lambda0  # Predicted average
    return norm(Y_avg_control_in_post - fit, 2) ** 2

# Step 7: Set constraints: lambda >= 0 and sum(lambda) = 1
lambda_constraints = [
    {'type': 'eq', 'fun': lambda l: np.sum(l[:-1]) - 1},  # Sum of lambda weights = 1
    {'type': 'ineq', 'fun': lambda l: l[:-1]}  # Lambda weights >= 0
]

# Step 8: Define initial guess (equal weights and zero intercept)
T_pre = Y_pre_control.shape[0]
initial_lambda_guess = np.zeros(T_pre + 1)
initial_lambda_guess[:-1] = 1 / T_pre  # Equal weights

# Step 9: Perform optimization
lambda_result = minimize(
    lambda_objective,
    initial_lambda_guess,
    constraints=lambda_constraints,
    bounds=[(0, 1)] * T_pre + [(None, None)]  # Bounds for lambda weights and no bounds for intercept
)

# Step 10: Extract optimized lambda weights and intercept
optimized_lambda_weights = lambda_result.x[:-1]
optimized_lambda_intercept = lambda_result.x[-1]

# Print results
print("Optimized Lambda Weights:", optimized_lambda_weights)
print("Optimized Intercept (Lambda_0):", optimized_lambda_intercept)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create the combined plot with lambda weights on the right y-axis
def plot_combined_with_right_weights(weights, pre_years, post_years, avg_control_Y, treatment_start):
    """
    Plot lambda weights and the average outcome Y for control units, with lambda weights on the right y-axis.

    Parameters:
    - weights: Optimized lambda weights (array-like, size T_pre)
    - pre_years: Pre-treatment years (array-like)
    - post_years: Post-treatment years (array-like)
    - avg_control_Y: Average outcome for control units over all years
    - treatment_start: The year when treatment starts
    """
    # Extend lambda weights to post-treatment with lambda_t = 1/T_post
    T_post = len(post_years)
    lambda_post = np.ones(T_post) / T_post  # Lambda_t = 1/T_post for all post-treatment years
    extended_weights = np.concatenate([weights, lambda_post])
    all_years = np.concatenate([pre_years, post_years])

    fig, ax1 = plt.subplots(figsize=(12, 6))

    # Plot average control outcome Y on the left y-axis
    avg_Y_over_time = data.groupby('year')['cigsale'].mean().values
    years = data['year'].unique()
    ax1.plot(years, avg_Y_over_time, marker='x', linestyle='--', color='green', label='Avg Control Units Y')
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Average Control Units Y', color='green')
    ax1.tick_params(axis='y', labelcolor='green')

    # Add a secondary y-axis for lambda weights on the right
    ax2 = ax1.twinx()
    ax2.plot(all_years, extended_weights, marker='o', linestyle='-', color='blue', label='Lambda Weights')
    ax2.set_ylabel('Lambda Weights', color='blue')
    ax2.tick_params(axis='y', labelcolor='blue')

    # Add vertical line for the treatment start year
    ax1.axvline(treatment_start, color='red', linestyle=':', linewidth=2, label=f'Treatment Start ({treatment_start})')

    # Add legends for both y-axes
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')

    # Add grid and title
    plt.title('Lambda Weights and Average Y for Control Units Over Time')
    plt.grid(True)
    plt.show()

# Generate the years for plotting
pre_treatment_years = pre_pivoted_data.index  # Pre-treatment years
post_treatment_years = post_period_data['year'].unique()  # Post-treatment years
treatment_start_year = 1989  # Treatment start year

# Call the function to plot
plot_combined_with_right_weights(optimized_lambda_weights, pre_treatment_years, post_treatment_years, Y_avg_control_in_post, treatment_start_year)
```

### SDID

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import minimize

data['Y_state_mean'] = data.groupby('state')['cigsale'].transform('mean')
data['Y_year_mean'] = data.groupby('year')['cigsale'].transform('mean')
data['Y_overall_mean'] = data['cigsale'].mean()
data['Y_demeaned'] = (
    data['cigsale'] - data['Y_state_mean'] - data['Y_year_mean'] + data['Y_overall_mean']
)

data['D'] = data['treated'] * data['post']
data['D_state_mean'] = data.groupby('state')['D'].transform('mean')
data['D_year_mean'] = data.groupby('year')['D'].transform('mean')
data['D_overall_mean'] = data['D'].mean()
data['D_demeaned'] = (
    data['D'] - data['D_state_mean'] - data['D_year_mean'] + data['D_overall_mean']
)

N_treated = len(data[data['treated'] == 1])
state_to_weight = dict(zip(pre_pivoted_data.columns[:-1], optimized_weights))
state_to_weight['California'] = 1 / N_treated

data['w'] = data['state'].map(state_to_weight).fillna(0)

T_pre = len(pre_period_data['year'].unique())
T_post = len(post_period_data['year'].unique())

lambda_weights = np.concatenate([optimized_lambda_weights, np.ones(T_post) / T_post])

data['lambda'] = data['year'].map(dict(zip(data['year'].unique(), lambda_weights)))

def treatment_effect_objective(tau_1):
    residuals = (
        data['Y_demeaned'] - tau_1 * data['D_demeaned']
    ) ** 2
    weighted_residuals = residuals * data['w'] * data['lambda']
    return weighted_residuals.sum()

initial_guess = [0]
result = minimize(treatment_effect_objective, initial_guess)
tau_1 = result.x[0]
print(tau_1)

all_states = list(state_to_weight.keys())
all_weights = list(state_to_weight.values())

plt.figure(figsize=(12, 6))
plt.bar(all_states, all_weights, color='skyblue', edgecolor='black')
plt.xlabel('States')
plt.ylabel('Weights')
plt.title('Weights for All States (Including California)')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
