# Survival Analysis

Survival analysis focuses on modeling time-to-event data, often used in clinical research to analyze patient outcomes.

## Naive Estimator

The naive survival estimator assumes that survival probabilities can be calculated directly from the observed data without accounting for censoring. It is defined as:

$$
S_{\text{naive}}(t) = \frac{\text{Number of individuals with } T > t}{N}
$$

Where:
- $T$ is the time-to-event variable.
- $N$ is the total number of individuals.

### Issues

The naive estimator does not account for **censored data**â€”cases where the event (e.g., death) has not occurred by the end of the observation period or where individuals drop out of the study. This results in biased survival estimates because:
1. Individuals who are censored are incorrectly treated as if they experienced the event.
2. Survival probabilities are underestimated when censoring is present, as the method assumes all censored individuals eventually experience the event.

---

## Kaplan-Meier Estimator

A more robust approach, such as the Kaplan-Meier estimator, adjusts for censoring by incorporating the survival probabilities only for individuals still at risk at each time point. The Kaplan-Meier estimator is a non-parametric statistic used to estimate the survival function, defined as:

$$
S(t) = P(T > t)
$$

Where:
- $T$ is the time-to-event variable.
- $t$ is the specific time point.

The Kaplan-Meier survival probability is computed iteratively as:

$$
\hat{S}(t) = \prod_{t_i \leq t} \left( 1 - \frac{d_i}{n_i} \right)
$$

Where:
- $d_i$: Number of events at time $t_i$.
- $n_i$: Number of individuals at risk just before $t_i$.

This approach adjusts for censored data, making it more reliable than the naive estimator.

---

## Cox Proportional Hazards Model

The Cox Proportional Hazards Model is a semi-parametric model used to relate covariates to survival times. The hazard function is expressed as:

$$
h(t \mid X) = h_0(t) \exp(\beta^\top X)
$$

Where:
- $h(t \mid X)$: Hazard function at time $t$ for covariates $X$.
- $h_0(t)$: Baseline hazard function.
- $\beta$: Vector of coefficients representing log-hazard ratios.

The coefficients $\beta$ are estimated by maximizing the partial likelihood:

$$
L(\beta) = \prod_{i=1}^n \frac{\exp(\beta^\top X_i)}{\sum_{j \in R_i} \exp(\beta^\top X_j)}
$$

Where:
- $R_i$: Risk set at the time of event for individual $i$.

---

## Concordance Index (C-Index)

The Concordance Index is used to evaluate the predictive accuracy of a survival model. It measures the proportion of correctly ordered pairs of survival times and is defined as:

$$
C = \frac{\text{Number of Concordant Pairs}}{\text{Total Number of Comparable Pairs}}
$$

A pair $(i, j)$ is concordant if:

$$
h(t_i \mid X_i) < h(t_j \mid X_j) \quad \text{and} \quad t_i < t_j
$$

or vice versa.

```python
import numpy as np
import pandas as pd
from sksurv.datasets import load_veterans_lung_cancer
from sksurv.nonparametric import kaplan_meier_estimator
from sksurv.linear_model import CoxPHSurvivalAnalysis
from sksurv.preprocessing import OneHotEncoder
from sksurv.metrics import concordance_index_censored
import matplotlib.pyplot as plt

# Load and preprocess the data
x, y = load_veterans_lung_cancer()
y_array = np.array(y.tolist())

# Naive Estimates
Tmax1 = int(y_array[:, 1].max())
Tmax2 = int(y_array[y_array[:, 0] == 1][:, 1].max())

N = len(y_array)
S1 = [np.sum(y_array[:, 1] > t) / N for t in range(Tmax1)]
S2 = [np.sum(y_array[y_array[:, 0] == 1][:, 1] > t) / N for t in range(Tmax2)]

# Kaplan-Meier Estimator
time_km, survival_prob_km = kaplan_meier_estimator(y_array[:, 0].astype(bool), y_array[:, 1])

# Survival by Treatment
treatment_curves = {}
for treatment_type in ("standard", "test"):
    mask = x["Treatment"] == treatment_type
    time_treatment, survival_prob_treatment = kaplan_meier_estimator(
        y_array[:, 0][mask].astype(bool), y_array[:, 1][mask]
    )
    treatment_curves[treatment_type] = (time_treatment, survival_prob_treatment)

# Survival by Cell Type
cell_curves = {}
for value in ("smallcell", "squamous", "adeno", "large"):
    mask = x["Celltype"] == value
    time_cell, survival_prob_cell = kaplan_meier_estimator(
        y_array[:, 0][mask].astype(bool), y_array[:, 1][mask]
    )
    cell_curves[value] = (time_cell, survival_prob_cell)

# Cox Proportional Hazards Model
x_encoded = OneHotEncoder().fit_transform(x)
cox_model = CoxPHSurvivalAnalysis()
cox_model.fit(x_encoded, y)
log_hazard_ratios = pd.Series(cox_model.coef_, index=x_encoded.columns)

# Predictions
x_new = pd.DataFrame.from_dict(
    {
        1: [65, 0, 0, 1, 60, 1, 0, 1],
        2: [65, 0, 0, 1, 60, 1, 0, 0],
        3: [65, 0, 1, 0, 60, 1, 0, 0],
        4: [65, 0, 1, 0, 60, 1, 0, 1],
    },
    columns=x_encoded.columns,
    orient="index",
)
pred_surv = cox_model.predict_survival_function(x_new)

# Evaluation
prediction = cox_model.predict(x_encoded)
concordance_index = concordance_index_censored(
    y["Status"], y["Survival_in_days"], prediction
)[0]

# Visualization: Grid of Graphs
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Naive Estimates Plot
axes[0, 0].plot(range(Tmax1), S1, label="Assuming dropouts died")
axes[0, 0].plot(range(Tmax2), S2, label="Assuming dropouts lived")
axes[0, 0].set_title("Naive Estimates")
axes[0, 0].legend()
axes[0, 0].set_xlabel("Time")
axes[0, 0].set_ylabel("Survival Probability")

# Kaplan-Meier Curve
axes[0, 1].step(time_km, survival_prob_km, where="post", label="KM Estimator")
axes[0, 1].plot(range(Tmax1), S1, label="Assuming dropouts died")
axes[0, 1].plot(range(Tmax2), S2, label="Assuming dropouts lived")
axes[0, 1].set_title("Kaplan-Meier Estimator")
axes[0, 1].legend()
axes[0, 1].set_xlabel("Time")
axes[0, 1].set_ylabel("Survival Probability")

# Survival by Treatment
for treatment_type, (time, survival_prob) in treatment_curves.items():
    axes[1, 0].step(time, survival_prob, where="post", label=f"Treatment = {treatment_type}")
axes[1, 0].set_title("Survival by Treatment")
axes[1, 0].legend()
axes[1, 0].set_xlabel("Time")
axes[1, 0].set_ylabel("Survival Probability")

# Survival by Cell Type
for cell_type, (time, survival_prob) in cell_curves.items():
    axes[1, 1].step(time, survival_prob, where="post", label=f"{cell_type}")
axes[1, 1].set_title("Survival by Cell Type")
axes[1, 1].legend()
axes[1, 1].set_xlabel("Time")
axes[1, 1].set_ylabel("Survival Probability")

plt.tight_layout()
plt.show()

# Compare Estimates: Summary Table
estimates_table = pd.DataFrame({
    "Estimator": ["Naive (Dropouts Died)", "Naive (Dropouts Lived)", "Kaplan-Meier"],
    "Survival at Tmax1": [
        S1[-1],
        S2[-1],
        survival_prob_km[-1]
    ]
})

print("Comparison of Survival Estimates")
print(estimates_table)

# Results Summary
results_summary = pd.DataFrame({
    "Covariate": log_hazard_ratios.index,
    "Log-Hazard Ratios": log_hazard_ratios.values,
    "Concordance Index": [concordance_index] * len(log_hazard_ratios),
})

print("\nCox Proportional Hazards Model Results")
print(results_summary)
```
