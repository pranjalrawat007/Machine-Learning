### Quantile Treatment Effects (QTE) with Controls

When estimating Quantile Treatment Effects (QTE), controls are included to account for confounding variables that may influence the outcome. These controls ensure the estimated treatment effects are not biased due to omitted variables.

#### Model with Controls

The quantile regression model with controls is:

$$ Q_y(\tau | X, \text{Treat}) = \alpha_\tau + \beta_{\text{treat}, \tau} \cdot \text{Treat} + \gamma_\tau^\top X $$

where:
- $\text{Treat}$ is the binary treatment variable (1 if treated, 0 otherwise).
- $X$ is a vector of control variables.
- $\alpha_\tau$ is the intercept specific to the $\tau^{th}$ quantile.
- $\beta_{\text{treat}, \tau}$ is the treatment effect at the $\tau^{th}$ quantile.
- $\gamma_\tau$ is the vector of coefficients for the control variables at the $\tau^{th}$ quantile.

#### Estimation

The estimation minimizes the weighted sum of residuals based on the quantile loss function:

$$ \min_{\beta, \gamma} \sum_{i=1}^n \rho_\tau \big( y_i - (\alpha_\tau + \beta_{\text{treat}, \tau} \cdot \text{Treat}_i + \gamma_\tau^\top X_i) \big) $$

where $\rho_\tau(u)$ is defined as:

$$ \rho_\tau(u) =
\begin{cases} 
\tau u & \text{if } u \geq 0 \\
(1-\tau) |u| & \text{if } u < 0
\end{cases}
$$

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
import warnings

# Suppress warnings
warnings.filterwarnings("ignore")

# Load Lalonde Dataset
url = "data/lalonde.csv"
data = pd.read_csv(url)

# Check if 'treat' exists
if 'treat' not in data.columns:
    raise ValueError("The 'treat' column is missing from the dataset.")

# Prepare Data
X = data[['age', 'educ', 'black', 'hispan', 'married', 'nodegree', 're74', 're75']]
X = sm.add_constant(X)  # Add constant for regression intercept
y = data["re78"]
treatment = data["treat"]

# Add treatment to predictors
X['treat'] = treatment

# Quantile Treatment Effect
quantiles = [0.1, 0.25, 0.5, 0.75, 0.9]
effects = []
lower_bounds = []
upper_bounds = []

for q in quantiles:
    try:
        model = sm.QuantReg(y, X).fit(q=q, max_iter=50000)  # Increase max_iter
        coef = model.params.get('treat', np.nan)
        conf_int = model.conf_int().loc['treat'] if 'treat' in model.conf_int().index else [np.nan, np.nan]
        effects.append(coef)
        lower_bounds.append(conf_int[0])
        upper_bounds.append(conf_int[1])
    except Exception as e:
        print(f"Error at quantile {q}: {e}")
        effects.append(np.nan)
        lower_bounds.append(np.nan)
        upper_bounds.append(np.nan)

# Plot Quantile Treatment Effects with Error Bands
plt.figure(figsize=(8, 5))
plt.errorbar(quantiles, effects, yerr=[np.subtract(effects, lower_bounds), np.subtract(upper_bounds, effects)],
             fmt='o', capsize=5, label="Treatment Effect")
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.title("Quantile Treatment Effects with Confidence Intervals")
plt.xlabel("Quantile")
plt.ylabel("Treatment Effect")
plt.grid()
plt.legend()
plt.show()
```

### Inverse Probability Weighting (IPW) for Quantile Treatment Effects

1. **Propensity Score (\( e(X) \)):**  
   Estimate the probability of receiving treatment using logistic regression:  
   $$
   \text{logit}(e(X)) = \alpha + \beta^\top X
   $$
   where \( e(X) = P(\text{Treat} = 1 \mid X) \).

2. **Weights (\( w_i \)):**  
   Assign inverse probability weights to balance the distribution of covariates between treated and control groups:  
   $$
   w_i =
   \begin{cases}
   \frac{1}{e(X_i)} & \text{if treated} \\
   \frac{1}{1 - e(X_i)} & \text{if not treated}
   \end{cases}
   $$

3. **Weighted Quantile Regression:**  
   Use the calculated weights in quantile regression to estimate the treatment effect at different quantiles (\( \tau \)) of the outcome distribution:
   $$
   Q_y(\tau \mid X, \text{Treat}) = \alpha_\tau + \beta_{\text{treat}, \tau} \cdot \text{Treat} + \gamma_\tau^\top X
   $$
   where:
   - \( \beta_{\text{treat}, \tau} \) is the treatment effect at the \( \tau \)-th quantile.

4. **Quantile Treatment Effect (\( \Delta_{\text{IPW}, \tau} \)):**  
   The difference in weighted conditional quantiles between treated and control groups:
   $$
   \Delta_{\text{IPW}, \tau} = Q_y(\tau \mid \text{Treat} = 1) - Q_y(\tau \mid \text{Treat} = 0)
   $$

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt

# Load Lalonde Dataset
url = "data/lalonde.csv"
data = pd.read_csv(url)

# Prepare Data
X = data[['age', 'educ', 'black', 'hispan', 'married', 'nodegree', 're74', 're75']]
X = sm.add_constant(X)  # Add constant for regression intercept
treatment = data["treat"]
y = data["re78"]

# Step 1: Estimate Propensity Scores
propensity_model = sm.Logit(treatment, X).fit(disp=False)
data['propensity_score'] = propensity_model.predict(X)

# Step 2: Compute IPW Weights
data['ipw'] = np.where(
    treatment == 1,
    1 / data['propensity_score'],
    1 / (1 - data['propensity_score'])
)

# Step 3: Weighted Quantile Treatment Effect Estimation
quantiles = [0.1, 0.25, 0.5, 0.75, 0.9]
effects = []
lower_bounds = []
upper_bounds = []

for q in quantiles:
    # Define weights for the treated and control groups
    weights = data['ipw']
    
    # Quantile regression with IPW weights
    model = sm.QuantReg(y, treatment).fit(q=q, weights=weights)
    coef = model.params.get('treat', np.nan)
    conf_int = model.conf_int().loc['treat'] if 'treat' in model.conf_int().index else [np.nan, np.nan]
    effects.append(coef)
    lower_bounds.append(conf_int[0])
    upper_bounds.append(conf_int[1])

# Step 4: Plot Quantile Treatment Effects with Confidence Intervals
plt.figure(figsize=(8, 5))
plt.errorbar(quantiles, effects, yerr=[np.subtract(effects, lower_bounds), np.subtract(upper_bounds, effects)],
             fmt='o', capsize=5, label="Treatment Effect")
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.title("Quantile Treatment Effects (IPW)")
plt.xlabel("Quantile")
plt.ylabel("Treatment Effect")
plt.grid()
plt.legend()
plt.show()
```

# Propensity Score Matching (PSM) for Quantile Treatment Effects

## Steps:

1. **Estimate Propensity Scores (\(e(X)\)):**
   - Use logistic regression to estimate the probability of treatment:
   $$
   \text{logit}(e(X)) = \alpha + \beta^\top X
   $$
   where \( e(X) = P(\text{Treat} = 1 \mid X) \).

2. **Match 1:1 Nearest Neighbors:**
   - For each treated unit, find the closest control unit based on the propensity score using nearest-neighbor matching.

3. **Plot Propensity Score Overlap:**
   - Compare the distributions of propensity scores before and after matching for treated and control groups to assess balance.

4. **Compute Quantile Treatment Effects (\( \Delta_{\text{PSM}, \tau} \)):**
   - For each quantile \( \tau \), compute the difference between the treated and control groups in the matched dataset:
   $$
   \Delta_{\text{PSM}, \tau} = Q_y(\tau \mid \text{Treat} = 1) - Q_y(\tau \mid \text{Treat} = 0)
   $$
   where \( Q_y(\tau) \) is the \( \tau \)-th quantile of the outcome \( Y \).

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors

# Load Lalonde Dataset
url = "data/lalonde.csv"
data = pd.read_csv(url)

# Prepare Data
X = data[['age', 'educ', 'black', 'hispan', 'married', 'nodegree', 're74', 're75']]
X = sm.add_constant(X)  # Add constant for regression intercept
treatment = data["treat"]
y = data["re78"]

# Step 1: Estimate Propensity Scores
propensity_model = sm.Logit(treatment, X).fit(disp=False)
data['propensity_score'] = propensity_model.predict(X)

# Step 2: Match 1:1 Nearest Neighbors
treated = data[data['treat'] == 1].reset_index()
control = data[data['treat'] == 0].reset_index()

# Use Nearest Neighbors to find matches
nn = NearestNeighbors(n_neighbors=1)
nn.fit(control[['propensity_score']])
distances, indices = nn.kneighbors(treated[['propensity_score']])

# Create a matched dataset
matched_indices = indices.flatten()
matched_control = control.iloc[matched_indices].reset_index(drop=True)
matched_data = pd.concat([treated, matched_control], axis=0).reset_index(drop=True)

# Step 3: Plot Propensity Score Overlap Pre- and Post-Matching
plt.figure(figsize=(12, 5))

# Pre-Matching
plt.subplot(1, 2, 1)
plt.hist(data[data['treat'] == 1]['propensity_score'], bins=20, alpha=0.6, label='Treated')
plt.hist(data[data['treat'] == 0]['propensity_score'], bins=20, alpha=0.6, label='Control')
plt.title("Propensity Score Distribution (Pre-Matching)")
plt.xlabel("Propensity Score")
plt.ylabel("Frequency")
plt.legend()

# Post-Matching
plt.subplot(1, 2, 2)
plt.hist(matched_data[matched_data['treat'] == 1]['propensity_score'], bins=20, alpha=0.6, label='Treated')
plt.hist(matched_data[matched_data['treat'] == 0]['propensity_score'], bins=20, alpha=0.6, label='Control')
plt.title("Propensity Score Distribution (Post-Matching)")
plt.xlabel("Propensity Score")
plt.ylabel("Frequency")
plt.legend()

plt.tight_layout()
plt.show()

# Step 4: Compute Quantile Treatment Effects
quantiles = [0.1, 0.25, 0.5, 0.75, 0.9]
effects = []
lower_bounds = []
upper_bounds = []

for q in quantiles:
    treated_y = matched_data[matched_data['treat'] == 1]['re78']
    control_y = matched_data[matched_data['treat'] == 0]['re78']
    
    # Compute quantiles
    treated_quantile = np.quantile(treated_y, q)
    control_quantile = np.quantile(control_y, q)
    
    # Compute treatment effect
    effect = treated_quantile - control_quantile
    effects.append(effect)
    # Placeholder for CI (can be computed with bootstrapping)
    lower_bounds.append(effect - 0.1)  # Example CI bounds
    upper_bounds.append(effect + 0.1)

# Plot Quantile Treatment Effects
plt.figure(figsize=(8, 5))
plt.errorbar(quantiles, effects, yerr=[np.subtract(effects, lower_bounds), np.subtract(upper_bounds, effects)],
             fmt='o', capsize=5, label="Treatment Effect")
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.title("Quantile Treatment Effects (PSM)")
plt.xlabel("Quantile")
plt.ylabel("Treatment Effect")
plt.grid()
plt.legend()
plt.show()
```
