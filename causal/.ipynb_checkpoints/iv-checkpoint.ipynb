# Instrumental Variables

Assumptions: 
- SUTVA
- Overlap
- Instrument Exogeneity
- Instrument Relevance
- Monotonicity (No defiers)
- Model Specification 

```python
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
np.random.seed(42)
n_simulations = 1000
n_samples = 500
gamma_true = 0.5
theta_true = 2.5
tau_true = 1

# Storage for estimators
beta_iv_estimators = []
beta_ols_estimators = []

# Simulate data
for _ in range(n_simulations):
    Z = np.random.uniform(0, 10, n_samples)
    u = np.random.normal(0, 1, n_samples)
    v = np.random.normal(0, 1, n_samples)

    X = gamma_true * Z + tau_true * u + v
    Y = theta_true * X + u

    # IV estimation
    beta_iv = np.cov(Z, Y)[0, 1] / np.cov(Z, X)[0, 1]
    beta_iv_estimators.append(beta_iv)

    # OLS estimation
    beta_ols = np.cov(X, Y)[0, 1] / np.var(X)
    beta_ols_estimators.append(beta_ols)

# Calculate theoretical standard error
sigma2 = np.var(Y - theta_true * X)
sst_x = np.sum((X - np.mean(X))**2)
rho2_xz = (np.corrcoef(Z, X)[0, 1])**2
stderr_theoretical = np.sqrt(sigma2 / (sst_x * rho2_xz))

# Mean and standard deviation of the simulated IV estimates
mean_iv = np.mean(beta_iv_estimators)
stderr_simulated = np.std(beta_iv_estimators)

# Plot histogram with theoretical standard error bounds
plt.figure(figsize=(10, 6))
plt.hist(beta_iv_estimators, bins=30, alpha=0.7, label="IV Estimator", color='blue', density=True)
plt.hist(beta_ols_estimators, bins=30, alpha=0.7, label="OLS Estimator", color='red', density=True)

# Adding lines for mean and 95% confidence interval
plt.axvline(x=mean_iv, color='black', linestyle='dashed', linewidth=2, label='Mean IV Estimate')
plt.axvline(x=mean_iv - 1.96 * stderr_theoretical, color='green', linestyle='dotted', linewidth=2, label='95% CI Lower')
plt.axvline(x=mean_iv + 1.96 * stderr_theoretical, color='green', linestyle='dotted', linewidth=2, label='95% CI Upper')

plt.title("Distribution of Estimators from Simulations (Model: X = gamma Z + tau * u + v, Y = theta * X + u)")
plt.xlabel("Estimator Value")
plt.ylabel("Density")
plt.legend()
plt.show()

# Display results
print(f"Mean IV Estimate (Simulation): {mean_iv:.4f}")
print(f"Simulated Standard Error: {stderr_simulated:.4f}")
print(f"Theoretical Standard Error: {stderr_theoretical:.4f}")
```

### 2SLS

```python
import numpy as np
import statsmodels.api as sm

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
n_samples = 50
n_bootstrap = 1000
gamma_true = 1
theta_true = 1
tau_true = 1

# Simulate data
Z = np.random.normal(0, 1, n_samples)
u = np.random.normal(0, 1, n_samples)
v = np.random.normal(0, 1, n_samples)

X = 2 + gamma_true * Z + tau_true * u + v
Y = 2 + theta_true * X + u

# Adding constant to the predictors
Z_with_const = sm.add_constant(Z)

# Step 1: Regress X on Z to get X_hat
model_X_on_Z = sm.OLS(X, Z_with_const).fit()
X_hat = model_X_on_Z.fittedvalues

# Step 2: Regress Y on X_hat
model_Y_on_X_hat = sm.OLS(Y, sm.add_constant(X_hat)).fit()
beta_hat = model_Y_on_X_hat.params[1]

# Define the Bootstrap function
def bootstrap(data_X_hat, data_Y, num_resamples=1000):
    bootstrap_betas = []
    n = len(data_X_hat)
    for _ in range(num_resamples):
        resample_indices = np.random.choice(n, size=n, replace=True)
        X_hat_resample = data_X_hat[resample_indices]
        Y_resample = data_Y[resample_indices]
        model_resample = sm.OLS(Y_resample, sm.add_constant(X_hat_resample)).fit()
        bootstrap_betas.append(model_resample.params[1])
    bootstrap_mean = np.mean(bootstrap_betas)
    bootstrap_stderr = np.std(bootstrap_betas, ddof=1)
    return bootstrap_mean, bootstrap_stderr

# Perform bootstrapping
bootstrap_mean, bootstrap_stderr = bootstrap(X_hat, Y, num_resamples=n_bootstrap)

# Calculate t-statistic
t_stat = beta_hat / bootstrap_stderr

# Display results
print(f"Estimate of Beta for X_hat: {beta_hat:.4f}")
print(f"Bootstrap Standard Error for Beta of X_hat: {bootstrap_stderr:.4f}")
print(f"T-statistic: {t_stat:.4f}")
```

```python
import numpy as np
import statsmodels.api as sm

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
n_samples = 500
n_bootstrap = 1000
n_vars = 5  # Number of x variables (x1 to xn)
n_instr = 5  # Number of instruments (z1 to zn)
tau_true = 2.0  # True beta for x1

# Simulate data for x1, x2, ..., xn, z1, ..., zn, and y
Z = np.random.normal(0, 1, (n_samples, n_instr))  # instruments z1 to zn
X = np.random.normal(0, 1, (n_samples, n_vars))  # exogenous variables x1 to xn
u = np.random.normal(0, 1, n_samples)
v = np.random.normal(0, 1, n_samples)

# Define the endogenous equation for x1
gamma = np.random.normal(0, 1, n_instr + n_vars - 1)  # coefficients for z1...zn, x2...xn
X[:, 0] = 2 + np.dot(X[:, 1:], gamma[n_instr:]) + np.dot(Z, gamma[:n_instr]) + u + v  # endogenize x1

# Define the outcome equation for y
theta = np.random.normal(0, 1, n_vars)  # coefficients for x1...xn
theta[0] = tau_true  # Set the true beta for x1
y = np.dot(X, theta) + u  # outcome variable y

# Adding constant to the predictors
Z_with_const = sm.add_constant(Z)
X_with_const = sm.add_constant(X)

# Step 1: Regress x1 (endogenous) on x2..xn and z1..zn to get X_hat
model_X_on_Z = sm.OLS(X[:, 0], np.column_stack([Z, X[:, 1:]])).fit()  # Include z1..zn, x2..xn
X_hat = model_X_on_Z.fittedvalues

# Step 2: Regress y on X_hat and other exogenous variables x2..xn
model_Y_on_X_hat = sm.OLS(y, sm.add_constant(np.column_stack([X_hat, X[:, 1:]]))).fit()

# Step 3: Bootstrap procedure to get standard error for the beta of X_hat
bootstrap_betas = []

# Perform bootstrapping
for _ in range(n_bootstrap):
    sample_indices = np.random.choice(n_samples, size=n_samples, replace=True)
    X_hat_bootstrap = X_hat[sample_indices]
    y_bootstrap = y[sample_indices]
    
    model_bootstrap = sm.OLS(y_bootstrap, sm.add_constant(np.column_stack([X_hat_bootstrap, X[sample_indices, 1:]]))).fit()
    bootstrap_betas.append(model_bootstrap.params[1])

# Compute the standard error from the bootstrap
stderr_bootstrap = np.std(bootstrap_betas, ddof=1)

# Estimate and t-statistic
beta_hat = model_Y_on_X_hat.params[1]
t_stat = beta_hat / stderr_bootstrap

# Display results
print(f"True Beta (tau_true): {tau_true:.4f}")
print(f"Estimate of Beta for X_hat: {beta_hat:.4f}")
print(f"Bootstrap Standard Error for Beta of X_hat: {stderr_bootstrap:.4f}")
print(f"T-statistic: {t_stat:.4f}")
```

### DML-IV

```python
import numpy as np
import lightgbm as lgb
import statsmodels.api as sm
from sklearn.model_selection import KFold

# Parameters
n_samples = 1000
n_vars = 2  # Number of x variables (x1 to xn)
n_instr = 2  # Number of instruments (z1 to zn)
true_tau = 2.0  # True beta for x1
n_folds = 10  # Number of folds for cross-validation
n_random_inits = 5  # Number of random initializations

results = []

for seed in range(n_random_inits):
    np.random.seed(seed)  # Set seed for each random initialization

    # Simulate data for x1, x2, ..., xn, z1, ..., zn, and y
    Z = np.random.normal(0, 1, (n_samples, n_instr))  # instruments z1 to zn
    X = np.random.normal(0, 1, (n_samples, n_vars))  # exogenous variables x1 to xn
    u = np.random.normal(0, 1, n_samples)
    v = np.random.normal(0, 1, n_samples)

    # Nonlinear function g for x1
    X[:, 0] = 0 * np.tanh(np.dot(X[:, 1:], np.random.normal(0, 1, n_vars - 1))) + \
              np.exp(np.abs(np.dot(Z, np.random.normal(0, 1, n_instr)))) + u + v

    # Nonlinear function f for y
    y = true_tau * X[:, 0] + np.sin(np.dot(X[:, 1:], np.random.normal(0, 1, n_vars - 1))) + u

    # Set up K-Fold cross-validation
    kf = KFold(n_splits=n_folds, shuffle=True, random_state=42)
    betas = []

    for train_index, test_index in kf.split(X):
        # Split data into training and test sets
        X_train, X_test = X[train_index], X[test_index]
        Z_train, Z_test = Z[train_index], Z[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # Combine X2..Xn and Z1..Zn
        train_features = np.column_stack([X_train[:, 1:], Z_train])
        test_features = np.column_stack([X_test[:, 1:], Z_test])

        # Train LightGBM for x1 residuals (r1_hat)
        model_x1 = lgb.LGBMRegressor(random_state=42, verbose=-1)
        model_x1.fit(train_features, X_train[:, 0])
        r1_hat_train = X_train[:, 0] - model_x1.predict(train_features)
        r1_hat_test = X_test[:, 0] - model_x1.predict(test_features)

        # Train LightGBM for y residuals (r2_hat)
        model_y = lgb.LGBMRegressor(random_state=42, verbose=-1)
        model_y.fit(train_features, y_train)
        r2_hat_train = y_train - model_y.predict(train_features)
        r2_hat_test = y_test - model_y.predict(test_features)

        # Regress r2_hat on r1_hat in the test fold
        model_ols = sm.OLS(r2_hat_test, sm.add_constant(r1_hat_test)).fit()
        betas.append(model_ols.params[1])

    # Average beta across folds for this initialization
    dml_iv_beta = np.mean(betas)
    results.append(dml_iv_beta)

# Display results
print(f"True Beta (tau_true): {true_tau:.4f}")
print(f"DML-IV Betas across {n_random_inits} random initializations: {results}")
print(f"Average DML-IV Beta: {np.mean(results):.4f}")
print(f"Standard Deviation of DML-IV Betas: {np.std(results):.4f}")
```

```python
import numpy as np
import lightgbm as lgb
import statsmodels.api as sm
from sklearn.model_selection import KFold

# Parameters
n_samples = 1000
n_vars = 2  # Number of x variables (x1 to xn)
n_instr = 2  # Number of instruments (z1 to zn)
true_tau = 2.0  # True beta for x1
n_folds = 10  # Number of folds for cross-validation
n_random_inits = 5  # Number of random initializations

results = []

for seed in range(n_random_inits):
    np.random.seed(seed)  # Set seed for each random initialization

    # Simulate data for x1, x2, ..., xn, z1, ..., zn, and y
    Z = np.random.normal(0, 1, (n_samples, n_instr))  # instruments z1 to zn
    X = np.random.normal(0, 1, (n_samples, n_vars))  # exogenous variables x1 to xn
    u = np.random.normal(0, 1, n_samples)
    v = np.random.normal(0, 1, n_samples)

    # Nonlinear function g for x1
    X[:, 0] = 0 * np.tanh(np.dot(X[:, 1:], np.random.normal(0, 1, n_vars - 1))) + \
              np.exp(np.abs(np.dot(Z, np.random.normal(0, 1, n_instr)))) + 0.1 * u + v

    # Nonlinear function f for y
    y = true_tau * X[:, 0] + np.sin(np.dot(X[:, 1:], np.random.normal(0, 1, n_vars - 1))) + u

    # Set up K-Fold cross-validation
    kf = KFold(n_splits=n_folds, shuffle=True, random_state=42)
    betas = []

    for train_index, test_index in kf.split(X):
        # Split data into training and test sets
        X_train, X_test = X[train_index], X[test_index]
        Z_train, Z_test = Z[train_index], Z[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # Combine X2..Xn and Z1..Zn
        train_features = np.column_stack([X_train[:, 1:], Z_train])
        test_features = np.column_stack([X_test[:, 1:], Z_test])

        # Train LightGBM for x1 residuals (r1_hat)
        model_x1 = lgb.LGBMRegressor(random_state=42, verbose=-1)
        model_x1.fit(train_features, X_train[:, 0])
        r1_hat_train = X_train[:, 0] - model_x1.predict(train_features)
        r1_hat_test = X_test[:, 0] - model_x1.predict(test_features)

        # Train LightGBM for y residuals (r2_hat)
        model_y = lgb.LGBMRegressor(random_state=42, verbose=-1)
        model_y.fit(train_features, y_train)
        r2_hat_train = y_train - model_y.predict(train_features)
        r2_hat_test = y_test - model_y.predict(test_features)

        # Regress r2_hat on r1_hat in the test fold
        model_ols = sm.OLS(r2_hat_test, sm.add_constant(r1_hat_test)).fit()
        betas.append(model_ols.params[1])

    # Average beta across folds for this initialization
    dml_iv_beta = np.mean(betas)
    results.append(dml_iv_beta)

# Display results
print(f"True Beta (tau_true): {true_tau:.4f}")
print(f"DML-IV Betas across {n_random_inits} random initializations: {results}")
print(f"Average DML-IV Beta: {np.mean(results):.4f}")
print(f"Standard Deviation of DML-IV Betas: {np.std(results):.4f}")
```
