# K-Means Clustering

K-Means Clustering partitions $n$ data points $\{x_1, x_2, \dots, x_n\}$ in $\mathbb{R}^d$ into $K$ clusters by minimizing the within-cluster sum of squares (WCSS). The objective function is:

$$
J = \sum_{k=1}^K \sum_{x_i \in C_k} \|x_i - \mu_k\|^2
$$

where:
- $C_k$ is the set of data points in cluster $k$,
- $\mu_k$ is the centroid of cluster $k$: $\mu_k = \frac{1}{|C_k|} \sum_{x_i \in C_k} x_i$,
- $\|x_i - \mu_k\|^2$ is the squared Euclidean distance between $x_i$ and $\mu_k$.

#### Algorithm

1. Initialization:
   - Select $K$ initial centroids $\mu_1, \mu_2, \dots, \mu_K$, either randomly or using the k-means++ method, which ensures centroids are well-separated.

2. Assignment Step:
   - Assign each data point $x_i$ to the cluster with the nearest centroid:
     $$
     C_i = \arg\min_{k} \|x_i - \mu_k\|^2
     $$
   This partitions the data into $K$ clusters.

3. Update Step:
   - Compute new centroids $\mu_k$ as the mean of all points assigned to cluster $k$:
     $$
     \mu_k = \frac{1}{|C_k|} \sum_{x_i \in C_k} x_i
     $$

4. Convergence:
   - Repeat the Assignment and Update steps until the centroids do not change significantly (e.g., based on a threshold or $\max \text{ iterations}$).

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate synthetic data
np.random.seed(42)
X = np.vstack((
    np.random.normal(loc=[2, 2], scale=0.8, size=(50, 2)),
    np.random.normal(loc=[6, 2], scale=0.8, size=(50, 2)),
    np.random.normal(loc=[4, 6], scale=0.8, size=(50, 2))
))

# Step 2: Initialize K-Means parameters
K = 3  # Number of clusters
initial_centroids = X[np.random.choice(X.shape[0], K, replace=False)]  # Random initialization
max_iters = 10
centroids_history = [initial_centroids.copy()]

# Step 3: K-Means Algorithm
for _ in range(max_iters):
    # Assignment Step
    distances = np.array([np.linalg.norm(X - centroid, axis=1) for centroid in centroids_history[-1]])
    assignments = np.argmin(distances, axis=0)

    # Update Step
    new_centroids = np.array([X[assignments == k].mean(axis=0) if len(X[assignments == k]) > 0 else centroids_history[-1][k] for k in range(K)])
    centroids_history.append(new_centroids)

# Step 4: Plotting the movement of centroids
plt.figure(figsize=(8, 8))
plt.scatter(X[:, 0], X[:, 1], s=30, c='lightgray', label='Data Points', alpha=0.6)

# Plot centroid movements
for k in range(K):
    trajectory = np.array([centroids[k] for centroids in centroids_history])
    plt.plot(trajectory[:, 0], trajectory[:, 1], marker='o', label=f'Centroid {k+1} Path')

# Final centroids
final_centroids = centroids_history[-1]
plt.scatter(final_centroids[:, 0], final_centroids[:, 1], s=100, c='red', label='Final Centroids')

plt.title("K-Means Centroid Movements")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# #### **Import Necessary Libraries**
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, accuracy_score
import warnings
warnings.filterwarnings('ignore')  # To suppress warnings for cleaner output

# Set a consistent style for plots
plt.style.use('ggplot')

# #### **Step 1: Generate Synthetic Dataset**
np.random.seed(42)
X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.60, random_state=42)

# Visualize the synthetic dataset
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
plt.title("Synthetic Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()

# #### **Step 2: Apply K-Means Clustering**
n_clusters = 4  # Number of clusters
kmeans = KMeans(n_clusters=n_clusters, init='k-means++', n_init=10, max_iter=300, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Visualize the clustering results
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', label='Data Points')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, label='Centroids')
plt.title(f"K-Means Clustering (K={n_clusters})")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#### **Step 3: Elbow Method to Determine Optimal K**
inertias = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=300, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(8, 6))
plt.plot(K_range, inertias, 'bo-', linewidth=2, markersize=8)
plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Sum of Squared Distances (Inertia)")
plt.xticks(K_range)
plt.grid(True)
plt.show()
```

#### Silhouette Score

The Silhouette Score measures the quality of clustering by comparing the cohesion (intra-cluster similarity) and separation (inter-cluster dissimilarity) of each point. For a point $x_i$, the silhouette score is defined as:

$$
s(x_i) = \frac{b(x_i) - a(x_i)}{\max(a(x_i), b(x_i))}
$$

where:
- $a(x_i)$: Average distance from $x_i$ to all other points in the same cluster.
  $$
  a(x_i) = \frac{1}{|C_k| - 1} \sum_{x_j \in C_k, j \neq i} \|x_i - x_j\|
  $$
- $b(x_i)$: Average distance from $x_i$ to all points in the nearest cluster (not containing $x_i$):
  $$
  b(x_i) = \min_{l \neq k} \frac{1}{|C_l|} \sum_{x_j \in C_l} \|x_i - x_j\|
  $$

The silhouette score for the entire dataset is the mean of individual scores:
$$
S = \frac{1}{n} \sum_{i=1}^n s(x_i)
$$

#### Properties of Silhouette Score
- $s(x_i)$ ranges from -1 to 1:
  - $s(x_i) \approx 1$: Well-clustered point, far from other clusters.
  - $s(x_i) \approx 0$: Point lies on the boundary between clusters.
  - $s(x_i) \approx -1$: Point assigned to the wrong cluster.
- Higher average silhouette scores indicate better clustering quality.

```python
# #### **Step 4: Silhouette Scores for Different K**
silhouette_scores = []

for k in K_range[1:]:  # Silhouette score is undefined for k=1
    kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=300, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

# Plot silhouette scores
plt.figure(figsize=(8, 6))
plt.plot(K_range[1:], silhouette_scores, 'go-', linewidth=2, markersize=8)
plt.title("Silhouette Scores for Different K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.xticks(K_range[1:])
plt.grid(True)
plt.show()
```

```python
# #### **Step 5: Compare Initialization Methods**
init_methods = ['random', 'k-means++']
plt.figure(figsize=(12, 5))

for i, method in enumerate(init_methods):
    kmeans = KMeans(n_clusters=n_clusters, init=method, n_init=10, max_iter=300, random_state=42)
    kmeans.fit(X)
    y_kmeans = kmeans.predict(X)
    centers = kmeans.cluster_centers_

    plt.subplot(1, 2, i + 1)
    plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', label='Data Points')
    plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, label='Centroids')
    plt.title(f"Initialization Method: {method}")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()

# #### **Step 6: Sensitivity to Number of Initializations (n_init)**
n_inits = [1, 5, 10, 20, 50]
inertias_n_init = []

for n in n_inits:
    kmeans = KMeans(n_clusters=n_clusters, init='random', n_init=n, max_iter=300, random_state=42)
    kmeans.fit(X)
    inertias_n_init.append(kmeans.inertia_)

# Plot inertias vs n_init
plt.figure(figsize=(8, 6))
plt.plot(n_inits, inertias_n_init, 'ms-', linewidth=2, markersize=8)
plt.title("Effect of n_init on Inertia")
plt.xlabel("Number of Initializations (n_init)")
plt.ylabel("Sum of Squared Distances (Inertia)")
plt.grid(True)
plt.show()
```

## Questions and Answers

**Question:** How does K-Means initialize centroids?  
**Answer:** Centroids can be initialized randomly or using the `'k-means++'` method, which selects initial cluster centers that are distant from each other to speed up convergence.

---

**Question:** What is the "Elbow Method"?  
**Answer:** The Elbow Method helps determine the optimal number of clusters by plotting the inertia against \( K \) and looking for the point where the rate of decrease sharply changes (the "elbow").

---

**Question:** What are the limitations of K-Means Clustering?  
**Answer:** K-Means assumes spherical clusters of similar size, is sensitive to outliers and initialization, and requires specifying the number of clusters \( K \) in advance.

---

**Question:** How does the `n_init` parameter affect the algorithm?  
**Answer:** `n_init` specifies how many times the algorithm will run with different centroid seeds. The final result is the best output in terms of inertia. Increasing `n_init` reduces the chance of getting stuck in a local minimum.

---

**Question:** What is inertia in K-Means Clustering?  
**Answer:** Inertia is the sum of squared distances of samples to their closest cluster center. It measures how internally coherent the clusters are.

---

**Question:** Can K-Means detect non-spherical clusters?  
**Answer:** No, K-Means performs poorly on datasets with non-spherical or unevenly sized clusters. Algorithms like DBSCAN or hierarchical clustering may perform better in such cases.

---

**Question:** How can you evaluate the quality of clustering?  
**Answer:** Use metrics like Silhouette Score, Calinski-Harabasz Index, or Davies-Bouldin Index, which assess the cohesion and separation of clusters.

---

**Question:** Why is K-Means sensitive to outliers?  
**Answer:** Outliers can distort the mean position of clusters because K-Means uses the mean to update centroids, leading to incorrect cluster assignments.

---

**Question:** How can you make K-Means more robust to outliers?  
**Answer:** Use algorithms like K-Medoids or DBSCAN, or preprocess the data to remove or reduce the influence of outliers.

---

**Question:** What is the effect of `max_iter`?  
**Answer:** `max_iter` sets the maximum number of iterations the algorithm will run for a single initialization. If convergence is not reached within this limit, the algorithm stops.

---

**Question:** How does the `tol` parameter affect convergence?  
**Answer:** `tol` defines the relative tolerance with regards to inertia to declare convergence. Smaller values lead to stricter convergence criteria.

---

**Question:** When should you choose K-Means Clustering?  
**Answer:** K-Means is suitable when you have a large dataset with spherical clusters of similar size and you know the number of clusters \( K \).

---

**Question:** How can you handle the need to specify \( K \) in K-Means?  
**Answer:** Use methods like the Elbow Method, Silhouette Analysis, or domain knowledge to estimate a reasonable value for \( K \).

### DBSCAN 

Density-Based Spatial Clustering of Applications with Noise (DBSCAN) clusters data points based on density, classifying points into core points, border points, and outliers. It groups points that are closely packed together and labels points in sparse regions as outliers.

Key Definitions

1. Neighborhood:
   The $\varepsilon$-neighborhood of a point $p$ is:
   $$
   N_{\varepsilon}(p) = \{q \in \mathbb{R}^d \mid \|p - q\| \leq \varepsilon\}
   $$

2. Core Point:
   A point $p$ is a core point if:
   $$
   |N_{\varepsilon}(p)| \geq \text{MinPts}
   $$

3. Border Point:
   A border point is within the $\varepsilon$-neighborhood of a core point but is not itself a core point.

4. Outlier (Noise Point):
   A point is an outlier if it is neither a core point nor a border point.

Algorithm Steps

1. Initialization:
   - Define the parameters:
     - $\varepsilon$: Radius of a point's neighborhood.
     - MinPts: Minimum number of neighbors for a point to qualify as a core point.
   - Mark all points as unvisited.

2. Iterate Over All Points:
   For each unvisited point $p$:
   - Compute its $\varepsilon$-neighborhood $N_{\varepsilon}(p)$.
   - If $p$ is a core point ($|N_{\varepsilon}(p)| \geq \text{MinPts}$):
     - Mark $p$ as visited.
     - Start a new cluster and assign $p$ to it.

3. Expand the Cluster:
   - For every point $q$ in $N_{\varepsilon}(p)$:
     - If $q$ is unvisited:
       - Mark $q$ as visited.
       - If $q$ is also a core point, add its neighbors $N_{\varepsilon}(q)$ to the cluster expansion process.
     - Add $q$ to the current cluster if not already assigned.

4. Classify Points:
   - Assign all density-reachable points (directly or indirectly connected to core points) to the same cluster.
   - Mark points that are not reachable from any core point as outliers (cluster label = $-1$).

5. Terminate:
   - Repeat steps 2–4 until all points are visited.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score

# Step 1: Generate synthetic data with outliers and varying densities
np.random.seed(42)
X, _ = make_blobs(n_samples=300, centers=[[0, 0], [3, 3], [-3, 3]], cluster_std=[0.5, 0.8, 0.3], random_state=42)
outliers = np.random.uniform(low=-6, high=6, size=(20, 2))  # Random outliers
X = np.vstack([X, outliers])

# Step 2: Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# Step 3: Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.8, min_samples=5)  # Parameters: neighborhood size and min neighbors
dbscan_labels = dbscan.fit_predict(X)

# Step 4: Visualize clustering results and highlight differences
fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# K-Means visualization
axes[0].scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', s=50, alpha=0.6, edgecolor='k')
axes[0].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='X', label='Centroids')
axes[0].set_title("K-Means Clustering (Disadvantages)")
axes[0].set_xlabel("Feature 1")
axes[0].set_ylabel("Feature 2")
axes[0].legend()
axes[0].text(-5.5, -5.5, "Disadvantages:\n- Forces all points into clusters\n- Sensitive to outliers\n- Assumes spherical clusters", fontsize=10, bbox=dict(facecolor='white', alpha=0.7))

# DBSCAN visualization
axes[1].scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis', s=50, alpha=0.6, edgecolor='k')
outliers_mask = dbscan_labels == -1
axes[1].scatter(X[outliers_mask, 0], X[outliers_mask, 1], color='black', s=50, label='Outliers', alpha=0.8)
axes[1].set_title("DBSCAN Clustering (Advantages)")
axes[1].set_xlabel("Feature 1")
axes[1].set_ylabel("Feature 2")
axes[1].legend()
axes[1].text(-5.5, -5.5, "Advantages:\n- Handles outliers effectively\n- Detects clusters of varying shapes/densities\n- No need to predefine number of clusters", fontsize=10, bbox=dict(facecolor='white', alpha=0.7))

plt.tight_layout()
plt.show()

# Step 5: Evaluate clustering
kmeans_silhouette = silhouette_score(X, kmeans_labels)
# Exclude outliers (-1 labels) for DBSCAN silhouette score
if len(set(dbscan_labels)) > 1:
    dbscan_silhouette = silhouette_score(X[dbscan_labels != -1], dbscan_labels[dbscan_labels != -1])
else:
    dbscan_silhouette = None

print(f"K-Means Silhouette Score: {kmeans_silhouette:.2f}")
if dbscan_silhouette is not None:
    print(f"DBSCAN Silhouette Score (excluding outliers): {dbscan_silhouette:.2f}")
else:
    print("DBSCAN Silhouette Score could not be computed (only one cluster detected).")
```

## Questions and Answers

Question: What is DBSCAN?  
Answer: DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a density-based clustering algorithm that groups points in dense regions and labels sparse points as outliers.

Question: What are the main parameters of DBSCAN?  
Answer: The two main parameters are:
- $\varepsilon$ (epsilon): The radius of a point's neighborhood.
- MinPts: The minimum number of points required in a neighborhood for a point to be considered a core point.

Question: How does DBSCAN handle outliers?  
Answer: DBSCAN labels points as outliers if they are not within the $\varepsilon$-neighborhood of any core point or cannot be reached by other density-reachable points.

Question: What is a core point in DBSCAN?  
Answer: A core point is a point that has at least MinPts neighbors within its $\varepsilon$-neighborhood:
$$
|N_{\varepsilon}(p)| \geq \text{MinPts}
$$

Question: What is a border point in DBSCAN?  
Answer: A border point lies within the $\varepsilon$-neighborhood of a core point but does not itself have enough neighbors to qualify as a core point.

Question: What is a noise point in DBSCAN?  
Answer: A noise point (outlier) is a point that is neither a core point nor density-reachable from any core point.

Question: What does density-reachable mean in DBSCAN?  
Answer: A point $q$ is density-reachable from a point $p$ if there exists a chain of core points $\{p_1, p_2, \dots, p_n\}$ such that:
$$
p_1 = p, \; p_n = q, \; \text{and } p_{i+1} \in N_{\varepsilon}(p_i), \; \forall i.
$$

Question: How does DBSCAN determine the number of clusters?  
Answer: DBSCAN does not require the number of clusters to be specified. Clusters are formed based on density, and the algorithm automatically determines the number of clusters.

Question: Can DBSCAN handle clusters of varying shapes?  
Answer: Yes, DBSCAN is effective at identifying clusters of arbitrary shapes, as it does not assume spherical cluster structures.

Question: What are the limitations of DBSCAN?  
Answer: DBSCAN has two main limitations:
1. It is sensitive to the choice of $\varepsilon$ and MinPts.
2. It struggles with datasets where clusters have widely varying densities.

Question: How can you choose the $\varepsilon$ parameter?  
Answer: A common approach is to use a k-distance graph (plotting the distance to the $k$th nearest neighbor) and look for the "elbow" in the graph, which indicates an appropriate $\varepsilon$.

Question: When should you use DBSCAN?  
Answer: DBSCAN is suitable for datasets with clusters of arbitrary shapes, where you expect outliers, and when the number of clusters is unknown or hard to estimate.

Question: How does DBSCAN compare to K-Means?  
Answer: Unlike K-Means:
- DBSCAN does not require the number of clusters to be specified.
- DBSCAN can identify outliers.
- DBSCAN can handle clusters of arbitrary shapes and varying densities, whereas K-Means assumes spherical clusters.
