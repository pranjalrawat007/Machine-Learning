# Imbalanced Classification

Generally these methods below only work well with primitive classifiers. They do not improve models like LightGBM. 

Methods: 
- Resampling Methods (oversampling, undersampling, SMOTE, Near Miss)
- Class weights
- Ensemble Methods (SMOTEBoost)

Evaluation Metrics: 
- Precision-Recall AUC
- TPR/FPR AUC

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, roc_curve, precision_recall_curve, auc
from lightgbm import LGBMClassifier


# Step 1: Create dataset generator
def create_dataset(
    n_samples=10000,
    weights=(0.99, 0.01),
    n_classes=2,
    class_sep=5,
    n_clusters=2,
):
    return make_classification(
        n_samples=n_samples,
        n_features=2,
        n_informative=2,
        n_redundant=0,
        n_repeated=0,
        hypercube=False,
        n_classes=n_classes,
        n_clusters_per_class=n_clusters,
        weights=list(weights),
        class_sep=class_sep,
        random_state=42,
    )

# Step 2: Generate the dataset
X, y = create_dataset(n_samples=10000, weights=(0.95, 0.05), n_classes=2)

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Step 3: Train LightGBM
model = LGBMClassifier(max_depth=3, random_state=42)
model.fit(X_train, y_train)

# Predict probabilities
y_train_pred_prob = model.predict_proba(X_train)[:, 1]
y_test_pred_prob = model.predict_proba(X_test)[:, 1]

# Calculate AUC for train and test
train_auc = roc_auc_score(y_train, y_train_pred_prob)
test_auc = roc_auc_score(y_test, y_test_pred_prob)

# Print AUC scores
print(f"Train AUC: {train_auc:.2f}")
print(f"Test AUC: {test_auc:.2f}")

# Step 4: Plot PR and ROC curves with threshold markers and annotations
def plot_curves_with_thresholds(y_true, y_pred_prob, title, ax_roc, ax_pr):
    # ROC Curve
    fpr, tpr, thresholds_roc = roc_curve(y_true, y_pred_prob)
    roc_auc = auc(fpr, tpr)
    ax_roc.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.2f})")
    ax_roc.plot([0, 1], [0, 1], linestyle='--', color='gray')
    ax_roc.set_title(f"ROC Curve - {title}")
    ax_roc.set_xlabel("False Positive Rate")
    ax_roc.set_ylabel("True Positive Rate")
    ax_roc.legend(loc="lower right")
    
    # Add vertical bars and annotate thresholds
    for i in range(0, len(thresholds_roc), max(1, len(thresholds_roc) // 10)):
        ax_roc.axvline(fpr[i], linestyle=":", color="red", alpha=0.6)
        ax_roc.text(fpr[i], tpr[i], f"{thresholds_roc[i]:.2f}", fontsize=8, color="red", rotation=45)

    # Precision-Recall Curve
    precision, recall, thresholds_pr = precision_recall_curve(y_true, y_pred_prob)
    pr_auc = auc(recall, precision)
    ax_pr.plot(recall, precision, label=f"PR Curve (AUC = {pr_auc:.2f})")
    ax_pr.set_title(f"Precision-Recall Curve - {title}")
    ax_pr.set_xlabel("Recall")
    ax_pr.set_ylabel("Precision")
    ax_pr.legend(loc="lower left")
    
    # Add vertical bars and annotate thresholds
    for i in range(0, len(thresholds_pr), max(1, len(thresholds_pr) // 10)):
        ax_pr.axvline(recall[i], linestyle=":", color="blue", alpha=0.6)
        ax_pr.text(recall[i], precision[i], f"{thresholds_pr[i]:.2f}", fontsize=8, color="blue", rotation=45)

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Train Curves
plot_curves_with_thresholds(y_train, y_train_pred_prob, "Train", axes[0, 0], axes[1, 0])

# Test Curves
plot_curves_with_thresholds(y_test, y_test_pred_prob, "Test", axes[0, 1], axes[1, 1])

plt.tight_layout()
plt.show()
```

### Resampling Methods

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from lightgbm import LGBMClassifier
from imblearn.over_sampling import RandomOverSampler, SMOTE, ADASYN
from imblearn.under_sampling import RandomUnderSampler, NearMiss
from sklearn.inspection import DecisionBoundaryDisplay

# Step 1: Create dataset generator
def create_dataset(
    n_samples=10000,
    weights=(0.99, 0.01),
    n_classes=2,
    class_sep=5,
    n_clusters=2,
):
    return make_classification(
        n_samples=n_samples,
        n_features=2,
        n_informative=2,
        n_redundant=0,
        n_repeated=0,
        hypercube=False,
        n_classes=n_classes,
        n_clusters_per_class=n_clusters,
        weights=list(weights),
        class_sep=class_sep,
        random_state=42,
    )

# Step 2: Generate the dataset
X, y = create_dataset(n_samples=10000, weights=(0.95, 0.05), n_classes=2)

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

from sklearn.metrics import precision_recall_curve, auc

# Step 3: Train and plot decision boundaries with PR AUC
def train_and_evaluate(X_train, y_train, X_test, y_test, sampler, title, ax):
    # Resample the training data if a sampler is provided
    if sampler is not None:
        X_train, y_train = sampler.fit_resample(X_train, y_train)
    
    # Train the model
    model = LGBMClassifier(max_depth=3, random_state=42)
    model.fit(X_train, y_train)
    
    # Predict probabilities for PR AUC calculation
    y_train_pred_prob = model.predict_proba(X_train)[:, 1]
    y_test_pred_prob = model.predict_proba(X_test)[:, 1]
    
    # Calculate PR AUC for train and test
    precision_train, recall_train, _ = precision_recall_curve(y_train, y_train_pred_prob)
    train_pr_auc = auc(recall_train, precision_train)
    precision_test, recall_test, _ = precision_recall_curve(y_test, y_test_pred_prob)
    test_pr_auc = auc(recall_test, precision_test)
    
    # Plot decision boundaries
    DecisionBoundaryDisplay.from_estimator(
        model, X_train, response_method="predict", alpha=0.5, ax=ax, cmap="coolwarm"
    )
    ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='viridis', edgecolor='k', alpha=0.7)
    ax.set_title(f"{title}\nTrain PR AUC: {train_pr_auc:.2f}, Test PR AUC: {test_pr_auc:.2f}")
    ax.set_xlabel("Feature 1")
    ax.set_ylabel("Feature 2")
    
    # Return metrics
    return train_pr_auc, test_pr_auc

# Step 4: Compare resampling methods
fig, axes = plt.subplots(3, 2, figsize=(15, 18))

results = []

# Original data (No resampling)
train_pr_auc, test_pr_auc = train_and_evaluate(X_train, y_train, X_test, y_test, None, "Original Data", axes[0, 0])
results.append(["Original", train_pr_auc, test_pr_auc])

# Naive Oversampling
train_pr_auc, test_pr_auc = train_and_evaluate(X_train, y_train, X_test, y_test, RandomOverSampler(random_state=42), "Naive Oversampling", axes[0, 1])
results.append(["Naive Oversampling", train_pr_auc, test_pr_auc])

# SMOTE
train_pr_auc, test_pr_auc = train_and_evaluate(X_train, y_train, X_test, y_test, SMOTE(random_state=42), "SMOTE", axes[1, 0])
results.append(["SMOTE", train_pr_auc, test_pr_auc])

# ADASYN
train_pr_auc, test_pr_auc = train_and_evaluate(X_train, y_train, X_test, y_test, ADASYN(random_state=42), "ADASYN", axes[1, 1])
results.append(["ADASYN", train_pr_auc, test_pr_auc])

# Random Under-Sampling
train_pr_auc, test_pr_auc = train_and_evaluate(X_train, y_train, X_test, y_test, RandomUnderSampler(random_state=42), "Random Under-Sampling", axes[2, 0])
results.append(["Random Under-Sampling", train_pr_auc, test_pr_auc])

# NearMiss
train_pr_auc, test_pr_auc = train_and_evaluate(X_train, y_train, X_test, y_test, NearMiss(version=1), "NearMiss", axes[2, 1])
results.append(["NearMiss", train_pr_auc, test_pr_auc])

plt.tight_layout()
plt.show()

# Step 5: Format results into a DataFrame
columns = [
    "Method", 
    "Train PR AUC",
    "Test PR AUC"
]
results_df = pd.DataFrame(results, columns=columns)

# Display the results
print(results_df)
```

### Class Weight Methods

```python
from sklearn.metrics import precision_recall_curve, auc

# Step 3: Train and plot decision boundaries with class weights
def train_and_evaluate_with_class_weights(X_train, y_train, X_test, y_test, class_weight, title, ax):
    # Train the model with specified class weights
    model = LGBMClassifier(max_depth=3, class_weight=class_weight, random_state=42)
    model.fit(X_train, y_train)
    
    # Predict probabilities for PR AUC calculation
    y_train_pred_prob = model.predict_proba(X_train)[:, 1]
    y_test_pred_prob = model.predict_proba(X_test)[:, 1]
    
    # Calculate PR AUC for train and test
    precision_train, recall_train, _ = precision_recall_curve(y_train, y_train_pred_prob)
    train_pr_auc = auc(recall_train, precision_train)
    precision_test, recall_test, _ = precision_recall_curve(y_test, y_test_pred_prob)
    test_pr_auc = auc(recall_test, precision_test)
    
    # Plot decision boundaries
    DecisionBoundaryDisplay.from_estimator(
        model, X_train, response_method="predict", alpha=0.5, ax=ax, cmap="coolwarm"
    )
    ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='viridis', edgecolor='k', alpha=0.7)
    ax.set_title(f"{title}\nTrain PR AUC: {train_pr_auc:.2f}, Test PR AUC: {test_pr_auc:.2f}")
    ax.set_xlabel("Feature 1")
    ax.set_ylabel("Feature 2")
    
    # Return metrics
    return train_pr_auc, test_pr_auc

# Step 4: Compare multiple class weight configurations
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

results = []

class_weights = [
    None,  # No class weights
    {0: 1, 1: 5},  # Weighting minority class higher
    {0: 1, 1: 10},  # Further increase in weight for minority class
    {0: 1, 1: 100},  # Extreme weighting for minority class
]

titles = [
    "No Class Weights",
    "Class Weights (0:1, 1:5)",
    "Class Weights (0:1, 1:10)",
    "Class Weights (0:1, 1:100)",
]

for class_weight, title, ax in zip(class_weights, titles, axes.ravel()):
    train_pr_auc, test_pr_auc = train_and_evaluate_with_class_weights(X_train, y_train, X_test, y_test, class_weight, title, ax)
    results.append([title, class_weight, train_pr_auc, test_pr_auc])

plt.tight_layout()
plt.show()

# Step 5: Format results into a DataFrame
columns = [
    "Configuration", 
    "Class Weights",
    "Train PR AUC",
    "Test PR AUC"
]
results_df = pd.DataFrame(results, columns=columns)

# Display the results
print(results_df)
```

### Ensemble Methods

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, average_precision_score
from imbens.ensemble import RUSBoostClassifier, SMOTEBoostClassifier, AsymBoostClassifier
from imbens.metrics import geometric_mean_score, make_index_balanced_accuracy
from lightgbm import LGBMClassifier

# Step 1: Generate an imbalanced dataset
RANDOM_STATE = 42
X, y = make_classification(
    n_classes=3,
    n_features=20,
    n_informative=15,
    n_redundant=5,
    weights=[0.1, 0.3, 0.6],
    flip_y=0,
    n_clusters_per_class=2,
    n_samples=5000,
    random_state=RANDOM_STATE,
)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.3, random_state=RANDOM_STATE
)

# Step 2: Define classifiers to evaluate
classifiers = {
    "RUSBoostClassifier": RUSBoostClassifier(n_estimators=50, random_state=RANDOM_STATE),
    "SMOTEBoostClassifier": SMOTEBoostClassifier(n_estimators=50, random_state=RANDOM_STATE),
    "AsymBoostClassifier": AsymBoostClassifier(n_estimators=50, random_state=RANDOM_STATE),
    "LightGBMClassifier": LGBMClassifier(
        n_estimators=50,
        class_weight="balanced",
        random_state=RANDOM_STATE, 
        verbose=-1
    ),
}

# Step 3: Evaluate classifiers
results = []
for name, clf in classifiers.items():
    #print(f"\nEvaluating {name}...")
    
    # Train the classifier
    clf.fit(X_train, y_train)
    
    # Predict on the test set
    y_pred = clf.predict(X_test)
    y_pred_proba = clf.predict_proba(X_test)
    
    # Metrics
    accuracy = accuracy_score(y_test, y_pred)
    geo_mean = geometric_mean_score(y_test, y_pred, average="macro")
    iba = make_index_balanced_accuracy(alpha=0.5, squared=True)(geometric_mean_score)(y_test, y_pred)
    
    # Calculate PR AUC for each class and average
    pr_auc = average_precision_score(
        pd.get_dummies(y_test), y_pred_proba, average="macro"
    )
    
    # Append results
    results.append({
        "Classifier": name,
        "Accuracy": accuracy,
        #"Geometric Mean": geo_mean,
        "Index Balanced Accuracy": iba,
        "PR AUC": pr_auc,
    })

# Step 4: Create a DataFrame for results
results_df = pd.DataFrame(results)

# Display the results
print("\nEvaluation Results:")
print(results_df)
```
