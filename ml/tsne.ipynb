# t-SNE

t-SNE (t-Distributed Stochastic Neighbor Embedding) is a dimensionality reduction algorithm designed for visualizing high-dimensional data in 2D or 3D. It works by preserving the local relationships between data points, making it excellent for clustering and identifying patterns. Unlike linear techniques like PCA, t-SNE captures complex, non-linear structures.

#### Algorithm

1. High-Dimensional Similarities:
   For data points $x_i$ and $x_j$, compute the probability $p_{ij}$ that $x_j$ is a neighbor of $x_i$:
   $$
   p_{ij} = \frac{\exp(-\|x_i - x_j\|^2 / 2\sigma_i^2)}{\sum_{k \neq i} \exp(-\|x_i - x_k\|^2 / 2\sigma_i^2)}
   $$
   - $\|x_i - x_j\|^2$: Squared Euclidean distance.
   - $\sigma_i$: Local scaling parameter based on perplexity.

2. Low-Dimensional Similarities:
   For mapped points $y_i$ and $y_j$, compute similarity using a t-distribution:
   $$
   q_{ij} = \frac{(1 + \|y_i - y_j\|^2)^{-1}}{\sum_{k \neq l} (1 + \|y_k - y_l\|^2)^{-1}}
   $$

3. KL Divergence Minimization:
   Minimize the mismatch between the high- and low-dimensional distributions $P$ and $Q$ using gradient descent:
   $$
   KL(P \| Q) = \sum_{i \neq j} p_{ij} \log\left(\frac{p_{ij}}{q_{ij}}\right)
   $$

4. Gradient Updates:
   Adjust the low-dimensional coordinates $y_i$ to minimize KL divergence:
   $$
   \Delta y_i = \eta \sum_{j} (p_{ij} - q_{ij})(y_i - y_j)(1 + \|y_i - y_j\|^2)^{-1}
   $$
   - $\eta$: Learning rate.

t-SNE can identify clusters in high-dimensional data (text, images, etc.) and find a non-linear dimensionality reduction. However, it does not preserve global relationships and is computationally expensive.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from mpl_toolkits.mplot3d import Axes3D

def compute_high_dim_similarities(X):
    n = X.shape[0]
    P = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            if i != j:
                P[i, j] = np.exp(-np.linalg.norm(X[i] - X[j])**2)
    P /= P.sum()
    return P

def compute_low_dim_similarities(Y):
    n = Y.shape[0]
    Q = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            if i != j:
                Q[i, j] = 1 / (1 + np.linalg.norm(Y[i] - Y[j])**2)
    Q /= Q.sum()
    return Q

def gradient_update(Y, P, Q, lr):
    n = Y.shape[0]
    gradients = np.zeros_like(Y)
    for i in range(n):
        for j in range(n):
            if i != j:
                diff = Y[i] - Y[j]
                gradients[i] += (P[i, j] - Q[i, j]) * diff / (1 + np.linalg.norm(diff)**2)
    Y += lr * gradients
    return Y

np.random.seed(42)
n_samples = 100
n_features = 3
n_clusters = 4
X, labels = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, cluster_std=1.5, random_state=42)
Y = np.random.rand(n_samples, 2)
Y_evolution = []
P = compute_high_dim_similarities(X)
learning_rate = 0.01
reduced_iterations = 1000

for iteration in range(reduced_iterations):
    print(iteration)
    Q = compute_low_dim_similarities(Y)
    Y = gradient_update(Y, P, Q, learning_rate)
    if iteration % 10 == 0:
        Y_evolution.append(Y.copy())

fig = plt.figure(figsize=(24, 8))

ax1 = fig.add_subplot(131, projection='3d')
ax1.scatter(X[:, 0], X[:, 1], X[:, 2], c=labels, cmap='viridis', s=50)
ax1.set_title("Original 3D Data")
ax1.set_xlabel("X1")
ax1.set_ylabel("X2")
ax1.set_zlabel("X3")

ax2 = fig.add_subplot(132)
for idx, Y_snapshot in enumerate(Y_evolution):
    ax2.scatter(Y_snapshot[:, 0], Y_snapshot[:, 1], c=labels, cmap='viridis', alpha=0.7, label=f"Iter {idx * 10}")
ax2.set_title("Evolution of Y (2D Mapping)")
ax2.set_xlabel("Y1")
ax2.set_ylabel("Y2")

ax3 = fig.add_subplot(133)
ax3.scatter(Y[:, 0], Y[:, 1], c=labels, cmap='viridis', s=50)
ax3.set_title("Final 2D Locations")
ax3.set_xlabel("Y1")
ax3.set_ylabel("Y2")

plt.tight_layout()
plt.show()
```

### Global vs Local Structure

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from sklearn import datasets, manifold

# Generate the S-curve dataset
n_samples = 1500
S_points, S_color = datasets.make_s_curve(n_samples, random_state=0)

# Define t-SNE parameters
n_components = 2  # number of coordinates for the manifold
perplexity = 30
n_iter = 250  # Corrected parameter name
random_state = 0

# Initialize t-SNE
t_sne = manifold.TSNE(
    n_components=n_components,
    perplexity=perplexity,
    init="random",
    n_iter=n_iter,  # Corrected to n_iter
    random_state=random_state,
)

# Apply t-SNE to the data
S_t_sne = t_sne.fit_transform(S_points)

# Plotting the original 3D structure and t-SNE results side by side
def plot_3d(points, points_color, title):
    x, y, z = points.T
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111, projection='3d')
    col = ax.scatter(x, y, z, c=points_color, s=50, alpha=0.8)
    ax.set_title(title)
    ax.set_xlabel("X1")
    ax.set_ylabel("X2")
    ax.set_zlabel("X3")
    fig.colorbar(col, ax=ax, orientation="horizontal", shrink=0.6, aspect=60, pad=0.01)

def plot_2d(points, points_color, title):
    fig, ax = plt.subplots(figsize=(6, 6), facecolor="white", constrained_layout=True)
    ax.scatter(points[:, 0], points[:, 1], c=points_color, s=50, alpha=0.8)
    ax.set_title(title)
    ax.xaxis.set_major_formatter(ticker.NullFormatter())
    ax.yaxis.set_major_formatter(ticker.NullFormatter())

# Plot original 3D data
plot_3d(S_points, S_color, "Original S-Curve in 3D")

# Plot t-SNE results
plot_2d(S_t_sne, S_color, "t-SNE 2D Projection")

plt.tight_layout()
plt.show()
```

### Questions and Answers

1. What are the key differences between t-SNE and PCA?  
t-SNE captures non-linear structures, focuses on local relationships, and is better for visualization. PCA is linear and preserves global variance.

2. What is perplexity in t-SNE?  
Perplexity controls the neighborhood size; lower values focus on local structures, while higher values emphasize global structures.

3. Why does t-SNE use a t-distribution in the low-dimensional space?  
The heavy-tailed t-distribution avoids the "crowding problem," where many points cluster too closely in lower dimensions.

4. What are the computational challenges of t-SNE?  
\( O(n^2) \) complexity due to pairwise similarity calculations. Barnes-Hut t-SNE and FIt-SNE reduce this to \( O(n \log n) \).

5. How would you validate or interpret t-SNE results?  
Use domain knowledge to assess clusters or compare with other algorithms like PCA or UMAP.

6. Can you explain how t-SNE would behave if applied to uniformly distributed data?  
t-SNE would struggle to find meaningful clusters and likely create artificial groupings.

7. Why might t-SNE fail to capture global structure, and how could you mitigate this?  
It focuses on local relationships. Mitigate by adjusting perplexity, using UMAP, or applying PCA before t-SNE.
