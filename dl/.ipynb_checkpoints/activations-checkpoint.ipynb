### ReLU

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# Parameters
num_layers = 20
layer_size = 10
input_size = 100
batch_size = 1

# Define Deep Network Class
class DeepNet(nn.Module):
    def __init__(self, num_layers, layer_size, input_size, activation_fn):
        super(DeepNet, self).__init__()
        layers = [nn.Linear(input_size, layer_size), activation_fn()]
        for _ in range(num_layers - 1):
            layers.append(nn.Linear(layer_size, layer_size))
            layers.append(activation_fn())
        self.network = nn.Sequential(*layers)

    def forward(self, x):
        return self.network(x)

# Function to measure gradient variance
def gradient_variance(model, init_scheme):
    for layer in model.modules():
        if isinstance(layer, nn.Linear):
            if init_scheme == "he_uniform":
                nn.init.kaiming_uniform_(layer.weight, nonlinearity='relu')
            elif init_scheme == "xavier_uniform":
                nn.init.xavier_uniform_(layer.weight)
            nn.init.constant_(layer.bias, 0)

    x = torch.randn(batch_size, input_size)
    target = torch.ones(batch_size, layer_size)

    # Forward and backward pass
    output = model(x)
    loss = nn.MSELoss()(output, target)
    loss.backward()

    # Collect gradient variances
    grad_variances = []
    for layer in model.modules():
        if isinstance(layer, nn.Linear):
            grad_variances.append(layer.weight.grad.var().item())

    return grad_variances

# Function to measure activation variance
def activation_variance(model, init_scheme):
    for layer in model.modules():
        if isinstance(layer, nn.Linear):
            if init_scheme == "he_uniform":
                nn.init.kaiming_uniform_(layer.weight, nonlinearity='relu')
            elif init_scheme == "xavier_uniform":
                nn.init.xavier_uniform_(layer.weight)
            nn.init.constant_(layer.bias, 0)

    x = torch.randn(batch_size, input_size)

    # Forward pass and collect activation variances
    activations = []
    for layer in model.network:
        x = layer(x)
        if isinstance(layer, (nn.ReLU, nn.Sigmoid)):
            activations.append(x.var().item())

    return activations

# Create models for comparison
relu_model = DeepNet(num_layers, layer_size, input_size, nn.ReLU)
sigmoid_model = DeepNet(num_layers, layer_size, input_size, nn.Sigmoid)

# Measure gradient variances
grad_var_relu = gradient_variance(relu_model, "he_uniform")
grad_var_sigmoid = gradient_variance(sigmoid_model, "xavier_uniform")

# Measure activation variances
act_var_relu = activation_variance(relu_model, "he_uniform")
act_var_sigmoid = activation_variance(sigmoid_model, "xavier_uniform")

# Plot gradient variance
plt.figure(figsize=(8, 5))
plt.plot(range(1, num_layers + 1), grad_var_relu, label="ReLU (He Init)", marker='o')
plt.plot(range(1, num_layers + 1), grad_var_sigmoid, label="Sigmoid (Xavier Init)", marker='x')
plt.xlabel("Layer Number")
plt.ylabel("Gradient Variance (Log Scale)")
plt.title("Gradient Variance: ReLU (He Init) vs Sigmoid (Xavier Init)")
plt.yscale('log')
plt.legend()
plt.grid()
plt.show()

# Plot activation variance
plt.figure(figsize=(8, 5))
plt.plot(range(1, num_layers + 1), act_var_relu, label="ReLU (He Init)", marker='o')
plt.plot(range(1, num_layers + 1), act_var_sigmoid, label="Sigmoid (Xavier Init)", marker='x')
plt.xlabel("Layer Number")
plt.ylabel("Activation Variance")
plt.title("Activation Variance: ReLU (He Init) vs Sigmoid (Xavier Init)")
plt.legend()
plt.grid()
plt.show()
```

### The Dying ReLU Problem

#### **ReLU Activation Function**
ReLU is defined as:
$$ f(x) = \max(0, x) $$

For inputs $x \leq 0$, the output and gradient are both $0$. This leads to the **dying ReLU problem**, where neurons stop contributing to learning because they no longer update during training.

---

#### **Numerical Example**

1. **Setup**:
   - Weight: $w = -0.5$, Bias: $b = -1.0$, Input: $x = 2.0$.
   - Pre-activation: 
     $$ z = wx + b = (-0.5)(2.0) + (-1.0) = -2.0 $$

2. **ReLU Activation**:
   $$ f(z) = \max(0, -2.0) = 0 $$

3. **Effect**:
   - Output: $0$
   - Gradient: $0$ (no weight updates).

---

### How Leaky ReLU Fixes This

Leaky ReLU introduces a small slope $\alpha$ for $z \leq 0$:
$$ f(z) = \begin{cases} z & \text{if } z > 0 \\ \alpha z & \text{if } z \leq 0 \end{cases} $$

For $\alpha = 0.1$:
$$ f(z) = 0.1 \times (-2.0) = -0.2 $$

---

#### **Comparison**:

| Activation Function | $z$  | Output       |
|----------------------|-------|--------------|
| ReLU                | -2.0 | $0$          |
| Leaky ReLU ($\alpha=0.1$) | -2.0 | $-0.2$ |

---

### Key Insight:
Leaky ReLU prevents neurons from "dying" by allowing a small gradient for negative inputs, enabling consistent learning.

```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

# Parameters
input_size = 1
output_size = 1
epochs = 5000  # Reduced epochs for visualization
lr = 0.01
iterations = 10  # Number of random initializations
loss_threshold = 0.1  # Threshold to determine convergence

# Generate a toy linear dataset
torch.manual_seed(42)
X = torch.linspace(-1, 1, 100).unsqueeze(1)  # 100 points between -1 and 1
true_w, true_b = 2, 1
y = true_w * X + true_b  # True linear relationship

# Define the regression model with configurable activation
class RegressionModel(nn.Module):
    def __init__(self, activation_fn):
        super(RegressionModel, self).__init__()
        self.linear = nn.Linear(input_size, output_size)
        self.activation = activation_fn()

    def forward(self, x):
        z = self.linear(x)
        y = self.activation(z)
        return y

# Function to train the model and return loss history
def train_model(activation_fn, alpha=None):
    if activation_fn == nn.LeakyReLU:
        model = RegressionModel(lambda: nn.LeakyReLU(negative_slope=alpha))
    else:
        model = RegressionModel(activation_fn)

    criterion = nn.MSELoss()
    optimizer = optim.SGD(model.parameters(), lr=lr)
    loss_history = []

    for epoch in range(epochs):
        optimizer.zero_grad()
        y_pred = model(X)
        loss = criterion(y_pred, y)
        loss.backward()
        optimizer.step()
        loss_history.append(loss.item())

    return loss_history

# Train models for ReLU and Leaky ReLU
relu_losses = []
leaky_relu_losses = []

for i in range(iterations):
    torch.manual_seed(i)  # Different initialization each time
    relu_loss = train_model(nn.ReLU)
    leaky_relu_loss = train_model(nn.LeakyReLU, alpha=0.1)
    relu_losses.append(relu_loss)
    leaky_relu_losses.append(leaky_relu_loss)

# Plot learning curves
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# ReLU Learning Curves
for i, loss_history in enumerate(relu_losses):
    axs[0].plot(range(epochs), loss_history, label=f"ReLU Init {i + 1}")
axs[0].set_title("ReLU: Learning Curves (Multiple Initializations)")
axs[0].set_xlabel("Epoch")
axs[0].set_ylabel("Loss")
axs[0].legend()
axs[0].grid()

# Leaky ReLU Learning Curves
for i, loss_history in enumerate(leaky_relu_losses):
    axs[1].plot(range(epochs), loss_history, label=f"Leaky ReLU Init {i + 1}")
axs[1].set_title("Leaky ReLU: Learning Curves (Multiple Initializations)")
axs[1].set_xlabel("Epoch")
axs[1].set_ylabel("Loss")
axs[1].legend()
axs[1].grid()

plt.tight_layout()
plt.show()
```
