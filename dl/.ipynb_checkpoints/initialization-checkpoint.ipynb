### Random vs Xavier Initialization

- Initialization needs to be random to break symmetry to let the network learn.
- But uniform initialization is an issue, because the variance of the activations shrinks (or explodes).
- This in turn causes the variance of the backproped gradients in the early layers to shrink (or explode).
- This makes learning very slow in deep sigmoid networks. Sigmoid network have this vanishing gradient problem.

#### Xavier Initialization
- Designed to maintain signal propagation through the network, ensuring consistent variance of inputs (and outputs) and gradients.
- Weights are drawn from a uniform distribution with bounds:
  $$ \text{Uniform}\left(-\sqrt{\frac{6}{n_\text{in} + n_\text{out}}}, \sqrt{\frac{6}{n_\text{in} + n_\text{out}}}\right) $$
  where $n_\text{in}$ and $n_\text{out}$ are the number of input and output units for the layer.

- This ensures the variance of the activations in each layer have similar variance to the inputs. 
- This mitigates (but cannot eliminate) the vanishing/exploding gradients.
- Keeps activation values in a stable range across layers, facilitating training of deep networks.

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# Parameters
num_layers = 10
layer_size = 20
input_size = 100
batch_size = 1
random_init_range = 0.05  # Range for random uniform initialization

# Define the deep sigmoid network
class DeepSigmoidNet(nn.Module):
    def __init__(self, num_layers, layer_size, input_size):
        super(DeepSigmoidNet, self).__init__()
        layers = [nn.Linear(input_size, layer_size), nn.Sigmoid()]
        for _ in range(num_layers - 1):
            layers.append(nn.Linear(layer_size, layer_size))
            layers.append(nn.Sigmoid())
        self.network = nn.Sequential(*layers)

    def forward(self, x):
        return self.network(x)

# Function to measure gradient variance
def gradient_variance(init_scheme):
    model = DeepSigmoidNet(num_layers, layer_size, input_size)
    if init_scheme == "random_uniform":
        for layer in model.modules():
            if isinstance(layer, nn.Linear):
                nn.init.uniform_(layer.weight, -random_init_range, random_init_range)
                nn.init.constant_(layer.bias, 0)
    elif init_scheme == "xavier_uniform":
        for layer in model.modules():
            if isinstance(layer, nn.Linear):
                nn.init.xavier_uniform_(layer.weight)
                nn.init.constant_(layer.bias, 0)

    x = torch.randn(batch_size, input_size)
    target = torch.ones(batch_size, layer_size)

    # Forward and backward pass
    output = model(x)
    loss = nn.MSELoss()(output, target)
    loss.backward()

    # Collect gradient variances
    grad_variances = []
    for layer in model.modules():
        if isinstance(layer, nn.Linear):
            grad_variances.append(layer.weight.grad.var().item())

    return grad_variances

# Calculate gradient variances
grad_var_random = gradient_variance("random_uniform")
grad_var_xavier = gradient_variance("xavier_uniform")

# Plot the results with log scale for y-axis
plt.figure(figsize=(8, 5))
plt.plot(range(1, num_layers + 1), grad_var_random, label="Random Uniform Init", marker='o')
plt.plot(range(1, num_layers + 1), grad_var_xavier, label="Xavier Uniform Init", marker='x')
plt.xlabel("Layer Number")
plt.ylabel("Gradient Variance (Log Scale)")
plt.title("Gradient Variance over Layers (Logarithmic Scale)")
plt.yscale('log')  # Logarithmic scale for gradient variance
plt.legend()
plt.grid()
plt.show()

# Function to measure activation variance
def activation_variance(init_scheme):
    model = DeepSigmoidNet(num_layers, layer_size, input_size)
    if init_scheme == "random_uniform":
        for layer in model.modules():
            if isinstance(layer, nn.Linear):
                nn.init.uniform_(layer.weight, -random_init_range, random_init_range)
                nn.init.constant_(layer.bias, 0)
    elif init_scheme == "xavier_uniform":
        for layer in model.modules():
            if isinstance(layer, nn.Linear):
                nn.init.xavier_uniform_(layer.weight)
                nn.init.constant_(layer.bias, 0)

    x = torch.randn(batch_size, input_size)

    # Forward pass and collect activation variances
    activations = []
    for layer in model.network:
        x = layer(x)
        if isinstance(layer, nn.Sigmoid):
            activations.append(x.var().item())

    return activations

# Calculate activation variances
act_var_random = activation_variance("random_uniform")
act_var_xavier = activation_variance("xavier_uniform")

# Plot activation variance
plt.figure(figsize=(8, 5))
plt.plot(range(1, num_layers + 1), act_var_random, label="Random Uniform Init", marker='o')
plt.plot(range(1, num_layers + 1), act_var_xavier, label="Xavier Uniform Init", marker='x')
plt.xlabel("Layer Number")
plt.ylabel("Activation Variance")
plt.title("Activation Variance over Layers")
plt.legend()
plt.grid()
plt.show()

```

### Xavier vs Kaiming Initialization

- Xavier initialization works well for sigmoid and tanh activations but struggles with ReLU networks.
- ReLU activations introduce asymmetry by zeroing out negative values, causing Xavier initialization to be suboptimal for deeper layers.
- In deep ReLU networks, this leads to diminishing or exploding variances of activations and gradients, slowing learning.

#### Kaiming Initialization
- Specifically designed for ReLU activations to address their unique properties.
- Weights are drawn from a uniform distribution with bounds:
  $$ \text{Uniform}\left(-\sqrt{\frac{6}{n_\text{in}}}, \sqrt{\frac{6}{n_\text{in}}}\right) $$
  where $n_\text{in}$ is the number of input units to the layer.
- Ensures consistent variance of activations across layers for ReLU networks.
- Maintains stable gradient flow and prevents the vanishing/exploding gradient problem, even in very deep networks.
- Kaiming initialization is superior to Xavier for ReLU-based architectures, making training faster and more stable.

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# Parameters
num_layers = 10
layer_size = 20
input_size = 100
batch_size = 1
random_init_range = 0.05  # Range for random uniform initialization

# Define the deep ReLU network
class DeepReLUNet(nn.Module):
    def __init__(self, num_layers, layer_size, input_size):
        super(DeepReLUNet, self).__init__()
        layers = [nn.Linear(input_size, layer_size), nn.ReLU()]
        for _ in range(num_layers - 1):
            layers.append(nn.Linear(layer_size, layer_size))
            layers.append(nn.ReLU())
        self.network = nn.Sequential(*layers)

    def forward(self, x):
        return self.network(x)

# Function to measure gradient variance
def gradient_variance(init_scheme):
    model = DeepReLUNet(num_layers, layer_size, input_size)
    if init_scheme == "random_uniform":
        for layer in model.modules():
            if isinstance(layer, nn.Linear):
                nn.init.uniform_(layer.weight, -random_init_range, random_init_range)
                nn.init.constant_(layer.bias, 0)
    elif init_scheme == "xavier_uniform":
        for layer in model.modules():
            if isinstance(layer, nn.Linear):
                nn.init.xavier_uniform_(layer.weight)
                nn.init.constant_(layer.bias, 0)
    elif init_scheme == "kaiming_uniform":
        for layer in model.modules():
            if isinstance(layer, nn.Linear):
                nn.init.kaiming_uniform_(layer.weight, nonlinearity='relu')
                nn.init.constant_(layer.bias, 0)

    x = torch.randn(batch_size, input_size)
    target = torch.ones(batch_size, layer_size)

    # Forward and backward pass
    output = model(x)
    loss = nn.MSELoss()(output, target)
    loss.backward()

    # Collect gradient variances
    grad_variances = []
    for layer in model.modules():
        if isinstance(layer, nn.Linear):
            grad_variances.append(layer.weight.grad.var().item())

    return grad_variances

# Function to measure activation variance
def activation_variance(init_scheme):
    model = DeepReLUNet(num_layers, layer_size, input_size)
    if init_scheme == "random_uniform":
        for layer in model.modules():
            if isinstance(layer, nn.Linear):
                nn.init.uniform_(layer.weight, -random_init_range, random_init_range)
                nn.init.constant_(layer.bias, 0)
    elif init_scheme == "xavier_uniform":
        for layer in model.modules():
            if isinstance(layer, nn.Linear):
                nn.init.xavier_uniform_(layer.weight)
                nn.init.constant_(layer.bias, 0)
    elif init_scheme == "kaiming_uniform":
        for layer in model.modules():
            if isinstance(layer, nn.Linear):
                nn.init.kaiming_uniform_(layer.weight, nonlinearity='relu')
                nn.init.constant_(layer.bias, 0)

    x = torch.randn(batch_size, input_size)

    # Forward pass and collect activation variances
    activations = []
    for layer in model.network:
        x = layer(x)
        if isinstance(layer, nn.ReLU):
            activations.append(x.var().item())

    return activations

# Calculate gradient and activation variances
grad_var_random = gradient_variance("random_uniform")
grad_var_xavier = gradient_variance("xavier_uniform")
grad_var_kaiming = gradient_variance("kaiming_uniform")

act_var_random = activation_variance("random_uniform")
act_var_xavier = activation_variance("xavier_uniform")
act_var_kaiming = activation_variance("kaiming_uniform")

# Plot gradient variance
plt.figure(figsize=(8, 5))
plt.plot(range(1, num_layers + 1), grad_var_random, label="Random Uniform Init", marker='o')
plt.plot(range(1, num_layers + 1), grad_var_xavier, label="Xavier Uniform Init", marker='x')
plt.plot(range(1, num_layers + 1), grad_var_kaiming, label="Kaiming Uniform Init", marker='s')
plt.xlabel("Layer Number")
plt.ylabel("Gradient Variance (Log Scale)")
plt.title("Gradient Variance over Layers (Logarithmic Scale)")
plt.yscale('log')  # Logarithmic scale for gradient variance
plt.legend()
plt.grid()
plt.show()

# Plot activation variance
plt.figure(figsize=(8, 5))
plt.plot(range(1, num_layers + 1), act_var_random, label="Random Uniform Init", marker='o')
plt.plot(range(1, num_layers + 1), act_var_xavier, label="Xavier Uniform Init", marker='x')
plt.plot(range(1, num_layers + 1), act_var_kaiming, label="Kaiming Uniform Init", marker='s')
plt.xlabel("Layer Number")
plt.ylabel("Activation Variance")
plt.title("Activation Variance over Layers")
plt.legend()
plt.grid()
plt.show()
```
