# Word2Vec

## CBOW Word Embedding Model with Brown Corpus

### 1. Data Preprocessing

Given a corpus $\mathcal{D}$ with sentences:

1. **Vocabulary Construction**:
   - Let $\mathcal{V}$ represent the vocabulary of words with a minimum frequency $k$.
   - Words appearing fewer than $k$ times are replaced with a special `<PAD>` token.

2. **Input Representation**:
   - Each sentence $s = [w_1, w_2, \ldots, w_n]$ is converted into a sequence of indices $s = [v(w_1), v(w_2), \ldots, v(w_n)]$, where $v(w)$ maps words to their corresponding vocabulary indices.

3. **Context-Target Pairs**:
   - For a center word $w_c$, its context words are defined as:
     $$
     \text{Context}(w_c) = [w_{c-w}, \ldots, w_{c-1}, w_{c+1}, \ldots, w_{c+w}]
     $$
     where $w$ is the window size.

### 2. CBOW Model Formulation

1. **Input**:
   - For a given target word $w_c$, the context words $[w_{c-w}, \ldots, w_{c+w}]$ are encoded as indices $[i_1, i_2, \ldots, i_{2w}]$.
   - These indices are mapped to dense vectors $[\mathbf{v}_{i_1}, \mathbf{v}_{i_2}, \ldots, \mathbf{v}_{i_{2w}}]$ using an embedding matrix $\mathbf{E} \in \mathbb{R}^{|\mathcal{V}| \times d}$, where $d$ is the embedding dimension.

2. **Averaging Context Vectors**:
   - The context embedding is computed by averaging:
     $$
     \mathbf{v}_{\text{context}} = \frac{1}{2w} \sum_{j=1}^{2w} \mathbf{v}_{i_j}
     $$

3. **Prediction**:
   - A linear transformation maps the averaged context vector to the vocabulary space:
     $$
     \mathbf{y} = \mathbf{W} \mathbf{v}_{\text{context}} + \mathbf{b}
     $$
     where $\mathbf{W} \in \mathbb{R}^{|\mathcal{V}| \times d}$ is the weight matrix, and $\mathbf{b} \in \mathbb{R}^{|\mathcal{V}|}$ is the bias vector.

4. **Softmax Output**:
   - The model predicts the target word by computing probabilities over the vocabulary:
     $$
     P(w_c | \text{context}) = \frac{\exp(\mathbf{y}[v(w_c)])}{\sum_{k=1}^{|\mathcal{V}|} \exp(\mathbf{y}[k])}
     $$

### 3. Training

1. **Loss Function**:
   - The model minimizes the negative log-likelihood of the target word:
     $$
     \mathcal{L} = -\sum_{(c, w_c) \in \mathcal{D}} \log P(w_c | \text{context}(w_c))
     $$

2. **Optimization**:
   - Parameters $\mathbf{E}, \mathbf{W}, \mathbf{b}$ are updated using the Adam optimizer.

### 4. Embedding Analysis

After training:

1. **Word Representations**:
   - Each word $w$ is represented by its embedding vector $\mathbf{v}_w$ from $\mathbf{E}$.

2. **Similarity**:
   - Cosine similarity between two words $w_1, w_2$ is defined as:
     $$
     \text{sim}(w_1, w_2) = \frac{\mathbf{v}_{w_1} \cdot \mathbf{v}_{w_2}}{\|\mathbf{v}_{w_1}\| \|\mathbf{v}_{w_2}\|}
     $$

3. **Querying Nearest Neighbors**:
   - To find the top $k$ most similar words to $w$, cosine similarity is computed for all vocabulary words, and the $k$ highest scores are selected.

```python
import torch
from torch import nn
from nltk.corpus import brown
from nltk import download
from collections import Counter
import random
from tqdm import tqdm

# Set random seed for reproducibility
torch.manual_seed(42)
random.seed(42)

# Load and Preprocess Data
print("Downloading and loading Brown Corpus...")
download('brown')
sentences = brown.sents()
print(f"Total sentences loaded: {len(sentences)}")

# Flatten sentences and count word frequencies
all_words = [word.lower() for sentence in sentences for word in sentence]
min_count = 5
word_counts = Counter(all_words)
print(f"Total unique words: {len(word_counts)}")

# Build vocabulary with continuous indices
filtered_words = [word for word, count in word_counts.items() if count >= min_count]
vocab = {word: idx + 1 for idx, word in enumerate(filtered_words)}
vocab["<PAD>"] = 0
num_special_tokens = 1  # Include <PAD>
print(f"Vocabulary size (including special tokens): {len(vocab)}")
print(f"Vocabulary size (excluding special tokens): {len(vocab) - num_special_tokens}")

# Convert sentences to indices and replace OOV words with <PAD>
data = [[vocab.get(word, vocab["<PAD>"]) for word in sentence] for sentence in sentences]
data = [sentence for sentence in data if len(sentence) > 2]
print(f"Filtered dataset size (sentences with rare words removed): {len(data)}")

# Reverse vocabulary for analysis
idx_to_word = {idx: word for word, idx in vocab.items()}

# Generate Context-Target Pairs
def generate_context_target_pairs(data, window_size=2):
    pairs = []
    for sentence in data:
        for center_idx in range(len(sentence)):
            context = []
            for w in range(-window_size, window_size + 1):
                if w == 0 or center_idx + w < 0 or center_idx + w >= len(sentence):
                    continue
                context.append(sentence[center_idx + w])
            while len(context) < 2 * window_size:  # Pad context to fixed length
                context.append(vocab["<PAD>"])
            pairs.append((context, sentence[center_idx]))
    return pairs

window_size = 2
pairs = generate_context_target_pairs(data, window_size)
print(f"Number of context-target pairs generated: {len(pairs)}")

# CBOW Model
class CBOWModel(nn.Module):
    def __init__(self, vocab_size, embedding_dim):
        super(CBOWModel, self).__init__()
        self.embeddings = nn.Embedding(vocab_size, embedding_dim, padding_idx=0)
        self.linear = nn.Linear(embedding_dim, vocab_size)

    def forward(self, context):
        embeds = self.embeddings(context)
        avg_embeds = embeds.mean(dim=1)
        out = self.linear(avg_embeds)
        return out

# Training Setup
embedding_dim = 50
learning_rate = 0.01
epochs = 20  # Increase the number of epochs for longer training
batch_size = 64

model = CBOWModel(len(vocab), embedding_dim)
print("\nModel Structure:\n", model)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

def get_batches(pairs, batch_size):
    random.shuffle(pairs)
    for i in range(0, len(pairs), batch_size):
        batch = pairs[i:i + batch_size]
        contexts, targets = zip(*batch)
        contexts_tensor = torch.tensor(contexts, dtype=torch.long)
        targets_tensor = torch.tensor(targets, dtype=torch.long)
        yield contexts_tensor, targets_tensor

# Training Loop with Progress Bar
print("Training started...")
for epoch in range(epochs):
    total_loss = 0
    batch_count = 0
    with tqdm(get_batches(pairs, batch_size), desc=f"Epoch {epoch + 1}", unit="batch") as progress_bar:
        for contexts, targets in progress_bar:
            optimizer.zero_grad()
            output = model(contexts)
            loss = criterion(output, targets)
            total_loss += loss.item()
            loss.backward()
            optimizer.step()
            batch_count += 1
            progress_bar.set_postfix(avg_loss=total_loss / batch_count)
    
    # Print top 5 similar words to 'government' at the end of each epoch
    if "government" in vocab:
        print(f"\nTop 5 similar words to 'government' at epoch {epoch + 1}:")
        print(find_similar("government"))

print("Training completed!")

# Analyze Word Vectors
word_vectors = model.embeddings.weight.data

def find_similar(word, top_k=5):
    if word not in vocab:
        return f"'{word}' is not in the vocabulary."
    word_idx = vocab[word]
    word_vector = word_vectors[word_idx]
    similarities = torch.matmul(word_vectors, word_vector) / (
        torch.norm(word_vectors, dim=1) * torch.norm(word_vector)
    )
    similarities[vocab["<PAD>"]] = -1.0  # Exclude <PAD> from results
    top_similar = torch.topk(similarities, top_k + 1).indices.tolist()
    return [idx_to_word[idx] for idx in top_similar if idx != word_idx][:top_k]

# Post-training examples
def post_training_examples(word_list, top_k=5):
    for word in word_list:
        if word in vocab:
            print(f"\nTop {top_k} similar words to '{word}':")
            print(find_similar(word, top_k=top_k))
        else:
            print(f"\n'{word}' not found in the vocabulary.")

# Analyze words after training
print("\nAnalyzing similar words after training:")
post_training_examples(["government", "country", "law", "city", "president"], top_k=5)

# Print sample embedding for 'government'
print("\nSample Embedding for 'government':")
if "government" in vocab:
    print(word_vectors[vocab["government"]])
else:
    print("'government' not in vocabulary.")
```

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import numpy as np

def visualize_embeddings(word_vectors, vocab, idx_to_word, num_words=200):
    # Get the first `num_words` from the vocabulary (ignoring <PAD>)
    words = list(vocab.keys())[:num_words]
    word_indices = [vocab[word] for word in words]
    
    # Extract corresponding embeddings
    embeddings = word_vectors[word_indices].cpu().numpy()

    # Perform t-SNE dimensionality reduction
    tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
    reduced_embeddings = tsne.fit_transform(embeddings)

    # Plot the embeddings
    plt.figure(figsize=(15, 15))
    for i, word in enumerate(words):
        x, y = reduced_embeddings[i]
        plt.scatter(x, y, marker='o', color='blue', alpha=0.7)
        plt.text(x + 0.1, y + 0.1, word, fontsize=9)

    plt.title("Word Embedding Visualization using t-SNE")
    plt.xlabel("Dimension 1")
    plt.ylabel("Dimension 2")
    plt.show()

# Visualize the embeddings for the top 200 words
visualize_embeddings(word_vectors, vocab, idx_to_word, num_words=200)
```
