# Normal Distributions

We demonstrate that Bayesian updating using a normal prior and likelihood yields the same posterior as directly combining two samples.

### Setup
- True mean: $\mu = 5$, true standard deviation: $\sigma = 2$
- Sample sizes: $n_1 = 30$, $n_2 = 40$

Samples are drawn from the same normal distribution.

### Step 1: Prior from Sample 1
The sample mean and variance for sample 1 are:
- Mean: $\bar{x}_1$
- Variance of mean: $\sigma_1^2 = \sigma^2 / n_1$

### Step 2: Posterior from Sample 2
The sample mean and variance for sample 2 are:
- Mean: $\bar{x}_2$
- Variance of mean: $\sigma_2^2 = \sigma^2 / n_2$

Using Bayesian updating:
- Posterior mean: $\mu_{\text{posterior}} = \frac{\bar{x}_1 / \sigma_1^2 + \bar{x}_2 / \sigma_2^2}{1 / \sigma_1^2 + 1 / \sigma_2^2}$
- Posterior variance: $\sigma_{\text{posterior}}^2 = \frac{1}{1 / \sigma_1^2 + 1 / \sigma_2^2}$

### Step 3: Direct Combination
Combining the two samples directly:
- Combined mean: $\bar{x}_{\text{combined}} = \frac{n_1 \bar{x}_1 + n_2 \bar{x}_2}{n_1 + n_2}$
- Combined variance: $\sigma_{\text{combined}}^2 = \sigma^2 / (n_1 + n_2)$

### Results
The posterior distribution from Bayesian updating matches the distribution obtained from direct combination of samples:
- Posterior mean = Combined mean = $4.731$
- Posterior standard deviation = Combined standard deviation = $0.239$

### Conclusion
Bayesian updating with a normal prior and likelihood is equivalent to directly combining samples when both are from the same normal distribution.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate two samples from the same normal distribution
np.random.seed(42)  # For reproducibility
true_mean = 5
true_std = 2
n1 = 30
n2 = 40

sample1 = np.random.normal(true_mean, true_std, n1)
sample2 = np.random.normal(true_mean, true_std, n2)

# Step 1: Compute prior from sample 1
mean1 = np.mean(sample1)
var1 = true_std**2 / n1  # Variance of the sample mean
std1 = np.sqrt(var1)

# Step 2: Use sample 2 to compute the posterior
mean2 = np.mean(sample2)
var2 = true_std**2 / n2  # Variance of the sample mean
std2 = np.sqrt(var2)

# Combine prior and sample 2 using Bayesian updating
posterior_mean = (mean1 / var1 + mean2 / var2) / (1 / var1 + 1 / var2)
posterior_var = 1 / (1 / var1 + 1 / var2)
posterior_std = np.sqrt(posterior_var)

# Step 3: Combine both samples directly
combined_sample = np.concatenate([sample1, sample2])
combined_mean = np.mean(combined_sample)
combined_var = true_std**2 / (n1 + n2)
combined_std = np.sqrt(combined_var)

# Visualization
x = np.linspace(true_mean - 3 * true_std, true_mean + 3 * true_std, 1000)

plt.figure(figsize=(12, 6))
plt.plot(x, 
         (1 / (std1 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean1) / std1)**2), 
         label="Prior (Sample 1)", linestyle="--")
plt.plot(x, 
         (1 / (std2 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean2) / std2)**2), 
         label="Likelihood (Sample 2)", linestyle="--")
plt.plot(x, 
         (1 / (posterior_std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - posterior_mean) / posterior_std)**2), 
         label="Posterior (Updated with Sample 2)")
plt.plot(x, 
         (1 / (combined_std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - combined_mean) / combined_std)**2), 
         label="Combined Sample (Direct)", linestyle=":")
plt.axvline(true_mean, color='red', linestyle=':', label='True Mean')
plt.title("Bayesian Updating vs. Direct Combination of Samples")
plt.xlabel("Mean")
plt.ylabel("Density")
plt.legend()
plt.show()

# Verify equivalence
posterior_mean, posterior_std, combined_mean, combined_std
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Parameters for the Beta prior
alpha_prior = 2  # Prior successes
beta_prior = 2   # Prior failures

# Parameters for the Normal prior
mean_prior = 0.5  # Initial belief about the coin bias
std_prior = 0.1   # Uncertainty about the prior belief

# Observations: Only heads
observed_heads = [1, 2, 3, 5, 10]
n_flips = observed_heads

# Possible values of coin bias
x = np.linspace(0, 1, 1000)

# Create a 2x1 grid for the plots
fig, axes = plt.subplots(2, 1, figsize=(10, 12))

# Plot 1: Beta prior/posterior
axes[0].set_title("Beta Prior and Posteriors")
for heads in observed_heads:
    alpha_post = alpha_prior + heads
    beta_post = beta_prior
    posterior = beta.pdf(x, alpha_post, beta_post)
    axes[0].plot(x, posterior, label=f"After {heads} heads")
# Plot the Beta prior
prior_beta = beta.pdf(x, alpha_prior, beta_prior)
axes[0].plot(x, prior_beta, label="Prior (Beta(2,2))", linestyle="--")
axes[0].set_xlabel("Coin Bias (p)")
axes[0].set_ylabel("Density")
axes[0].legend()

# Plot 2: Normal prior/posterior
axes[1].set_title("Normal Prior and Posteriors")
mean_post = mean_prior
std_post = std_prior
for heads in observed_heads:
    # Compute the posterior for Normal prior
    mean_post = (mean_post / std_post**2 + heads / (std_prior**2 * n_flips[-1])) / (1 / std_post**2 + n_flips[-1] / std_prior**2)
    std_post = np.sqrt(1 / (1 / std_post**2 + n_flips[-1] / std_prior**2))
    posterior_normal = (1 / (std_post * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean_post) / std_post)**2)
    axes[1].plot(x, posterior_normal, label=f"After {heads} heads")
# Plot the Normal prior
prior_normal = (1 / (std_prior * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean_prior) / std_prior**2))
axes[1].plot(x, prior_normal, label="Prior (Normal(0.5, 0.1))", linestyle="--")
axes[1].set_xlabel("Coin Bias (p)")
axes[1].set_ylabel("Density")
axes[1].legend()

plt.tight_layout()
plt.show()
```
