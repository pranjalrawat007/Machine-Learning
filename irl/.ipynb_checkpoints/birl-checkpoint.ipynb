### Bayesian Inverse Reinforcement Learning (Bayesian IRL)

#### Problem Setup
The objective is to recover the reward function $R(s)$ in a Markov Decision Process (MDP) such that the expert's demonstrated trajectories are most likely under a stochastic policy derived from the recovered rewards.

---

#### Transition Model
For an MDP with:
- $S$: number of states,
- $A$: number of actions,
- $P(s'|s,a)$: transition probabilities,
- $\gamma$: discount factor,

the expert's trajectories specify state-action pairs $(s, a)$ sampled from the expert's policy.

---

#### Reward Function
The reward function is represented as:
$$
R(s) \in [-R_{\text{max}}, R_{\text{max}}].
$$

---

#### Expert Policy
The expert policy $\pi^*(a|s)$ is assumed to be optimal under the expert's reward function $R(s)$, with the probability of taking action $a$ in state $s$ proportional to:
$$
\pi^*(a|s) \propto \exp\left(\alpha Q(s, a)\right),
$$
where:
- $Q(s, a)$ is the action-value function,
- $\alpha$ is a precision parameter modeling the expert's confidence in optimal actions.

---

#### Action-Value Function
The $Q$-value for a state-action pair is:
$$
Q(s, a) = R(s) + \gamma \sum_{s'} P(s'|s,a) V(s'),
$$
where the value function $V(s)$ satisfies:
$$
V(s) = \max_a Q(s, a).
$$

---

#### Log-Likelihood of Expert Trajectories
The log-likelihood of the expert's trajectories $\mathcal{D}$ under the reward function $R(s)$ is:
$$
\mathcal{L}(R) = \sum_{(s, a) \in \mathcal{D}} \left[\alpha Q(s, a) - \log \sum_{a'} \exp(\alpha Q(s, a'))\right].
$$

---

#### Bayesian Framework
1. **Prior**:
   A uniform prior is assumed over $R(s)$:
   $$ 
   p(R) \propto 1.
   $$

2. **Posterior**:
   The posterior probability of $R(s)$ given the expert's trajectories is proportional to:
   $$
   p(R|\mathcal{D}) \propto \exp(\mathcal{L}(R)) \cdot p(R).
   $$

---

#### Sampling via MCMC
Bayesian IRL uses Markov Chain Monte Carlo (MCMC) to sample from the posterior distribution of $R(s)$.

1. **Proposal**:
   Generate a candidate reward function $R'(s)$ by perturbing a randomly chosen state's reward:
   $$
   R'(s_i) = R(s_i) + \text{step size}, \quad R'(s_i) \in [-R_{\text{max}}, R_{\text{max}}].
   $$

2. **Acceptance Criterion**:
   Accept $R'(s)$ with probability:
   $$
   \min\left(1, \exp\left(\mathcal{L}(R') - \mathcal{L}(R)\right)\right).
   $$

---

#### Algorithm Steps
1. Initialize $R(s)$ randomly in $[-R_{\text{max}}, R_{\text{max}}]$.
2. For each MCMC iteration:
   - Propose a new reward function $R'(s)$.
   - Compute the posterior probability $\exp(\mathcal{L}(R'))$.
   - Accept or reject $R'(s)$ based on the acceptance criterion.
3. Collect samples after a burn-in period.

```python
import numpy as np
import matplotlib.pyplot as plt
import copy

# Set random seed for reproducibility
np.random.seed(42)

###################################################################
# Simple GridWorld Setup
# We use a 4x4 grid:
# States: 0 to 15 (row-major)
# Actions: 0=Left, 1=Down, 2=Right, 3=Up
# Terminal state at bottom-right corner (state 15)
#
# Transition dynamics:
# - Move in chosen direction if possible, else stay.
# - Episode ends if we reach terminal state.
###################################################################

N = 4
NUM_STATES = N * N
NUM_ACTIONS = 4
GAMMA = 0.9
ALPHA = 10.0
STEP_SIZE = 0.1
R_MAX = 1.0
N_SAMPLES = 500

def state_to_rc(s):
    return (s // N, s % N)

def rc_to_state(r,c):
    return r*N + c

def step_func(s, a):
    if s == (NUM_STATES - 1):
        return s, 0, True
    r,c = state_to_rc(s)
    if a == 0 and c > 0:   # left
        c -= 1
    elif a == 1 and r < N-1: # down
        r += 1
    elif a == 2 and c < N-1: # right
        c += 1
    elif a == 3 and r > 0:   # up
        r -= 1
    ns = rc_to_state(r,c)
    done = (ns == (NUM_STATES - 1))
    return ns, 0, done

# Precompute transitions
P = {s: {a: [] for a in range(NUM_ACTIONS)} for s in range(NUM_STATES)}
for s in range(NUM_STATES):
    for a in range(NUM_ACTIONS):
        ns, _, done = step_func(s,a)
        P[s][a].append((1.0, ns, None, done))


###################################################################
# DP functions for a given reward function
###################################################################
def policy_evaluation(pi, rewards, gamma=GAMMA, theta=1e-5):
    V = np.zeros(NUM_STATES)
    while True:
        delta = 0
        for s in range(NUM_STATES):
            v = 0
            for a in range(NUM_ACTIONS):
                p = pi[s,a]
                for p_sa, ns, _, done in P[s][a]:
                    r = rewards[s] 
                    v += p * p_sa * (r + gamma*(0 if done else V[ns]))
            delta = max(delta, abs(v-V[s]))
            V[s] = v
        if delta < theta:
            break
    return V

def policy_improvement(V, rewards, gamma=GAMMA):
    Q = np.zeros((NUM_STATES, NUM_ACTIONS))
    for s in range(NUM_STATES):
        for a in range(NUM_ACTIONS):
            for p_sa, ns, _, done in P[s][a]:
                r = rewards[s]
                Q[s,a] += p_sa*(r + gamma*(0 if done else V[ns]))
    pi = np.zeros((NUM_STATES, NUM_ACTIONS))
    best_actions = np.argmax(Q, axis=1)
    for s in range(NUM_STATES):
        pi[s, best_actions[s]] = 1.0
    return Q, pi

def policy_iteration(rewards):
    pi = np.ones((NUM_STATES, NUM_ACTIONS))/NUM_ACTIONS
    while True:
        V = policy_evaluation(pi, rewards)
        old_pi = pi.copy()
        Q, pi = policy_improvement(V, rewards)
        if np.all(pi == old_pi):
            return Q, pi

###################################################################
# Expert demonstration
# The expert policy:
# Move RIGHT until last column, then move DOWN until goal.
###################################################################
expert_trajectory = [(0,2),(1,2),(2,2),(3,1),(7,1),(11,1),(15,None)]
expert_pairs = expert_trajectory[:-1]

###################################################################
# Bayesian IRL
###################################################################
def log_likelihood(rewards):
    Q, pi = policy_iteration(rewards)
    ll = 0.0
    for (s,a) in expert_pairs:
        logits = ALPHA*Q[s,:]
        lse = np.log(np.sum(np.exp(logits)))
        ll += (ALPHA*Q[s,a] - lse)
    return ll

def prior_log_prob(rewards):
    # Uniform prior over [-R_MAX, R_MAX] is constant => log(prob) = const => 0
    return 0.0

def posterior_log_prob(rewards):
    return log_likelihood(rewards) + prior_log_prob(rewards)

def propose_new_rewards(rewards):
    new_rewards = rewards.copy()
    idx = np.random.randint(NUM_STATES)
    new_rewards[idx] += np.random.choice([-STEP_SIZE, STEP_SIZE])
    new_rewards[idx] = np.clip(new_rewards[idx], -R_MAX, R_MAX)
    return new_rewards

# MCMC sampling
rewards = np.random.uniform(-R_MAX, R_MAX, NUM_STATES)
logp = posterior_log_prob(rewards)

samples = []
accepts = 0
for i in range(N_SAMPLES):
    candidate = propose_new_rewards(rewards)
    logp_cand = posterior_log_prob(candidate)
    ratio = np.exp(logp_cand - logp)
    if np.random.rand() < ratio:
        rewards = candidate
        logp = logp_cand
        accepts += 1
    samples.append(rewards.copy())
print("Acceptance rate:", accepts/N_SAMPLES)

# Recovered rewards
recovered_rewards = np.mean(samples[int(N_SAMPLES/2):], axis=0)
recovered_rewards_2d = recovered_rewards.reshape(N,N)


###################################################################
# Plotting the recovered rewards
###################################################################
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.title("Sampled Rewards (final sample)")
plt.imshow(rewards.reshape(N,N), cmap='Spectral', vmin=-1, vmax=1)
plt.colorbar()

plt.subplot(1,2,2)
plt.title("Recovered Rewards (mean last half)")
plt.imshow(recovered_rewards_2d, cmap='Spectral', vmin=-1, vmax=1)
plt.colorbar()
plt.tight_layout()
plt.show()

###################################################################
# Show policies as arrows
# Expert policy: from each non-terminal state:
# if c < N-1: action=RIGHT
# else if r < N-1: action=DOWN
# else terminal
###################################################################
def action_to_vector(a):
    # Actions: 0=Left,1=Down,2=Right,3=Up
    if a == 0: return (-1,0)
    if a == 1: return (0,1)
    if a == 2: return (1,0)
    if a == 3: return (0,-1)

def get_expert_policy():
    pi_exp = np.zeros((NUM_STATES, NUM_ACTIONS))
    for s in range(NUM_STATES):
        if s == NUM_STATES - 1:
            pi_exp[s,:] = 0
            continue
        r,c = state_to_rc(s)
        if c < N-1:
            a = 2 # right
        elif r < N-1:
            a = 1 # down
        else:
            a = 0 # doesn't matter for terminal but let's say left
        pi_exp[s,:] = 0
        pi_exp[s,a] = 1.0
    return pi_exp

pi_exp = get_expert_policy()
# Get policy from recovered rewards
_, pi_rec = policy_iteration(recovered_rewards)

# Plotting policies as arrows
def plot_policy_arrows(pi, title):
    plt.title(title)
    X, Y, U, V = [], [], [], []
    for s in range(NUM_STATES):
        if s == NUM_STATES-1:
            continue
        a = np.argmax(pi[s])
        dx, dy = action_to_vector(a)
        r,c = state_to_rc(s)
        # Note: Matplotlib's origin is top-left. We'll invert y for display
        X.append(c)
        Y.append(r)
        U.append(dx)
        V.append(-dy)  # invert Y direction for plotting so up is up
    plt.xlim(-0.5,N-0.5)
    plt.ylim(-0.5,N-0.5)
    plt.gca().invert_yaxis()
    plt.xticks(range(N))
    plt.yticks(range(N))
    plt.grid(True)
    plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1, color='blue')

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plot_policy_arrows(pi_exp, "Expert Policy")

plt.subplot(1,2,2)
plot_policy_arrows(pi_rec, "Recovered Policy")

plt.tight_layout()
plt.show()
```
